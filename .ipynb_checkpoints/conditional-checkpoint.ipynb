### Comparision operators 'is' and 'in'

- **in** : Evaluates to True when the object being compared is inside iterable.
- **is** : Evaluates to True when two different names are refering to a same object in memory.

```python
""" Overview of 'is' and 'in' """

def master_comparison_with_nuances():
    print("=== PART 1: The 'is' Operator (Identity) ===\n")
    
    # 1. Comparison with None (Best Practice)
    status = None
    if status is None:
        print("1. Status is None (Correct check)")

    # 2. String Interning (Literals - The "Compile Time" Logic)
    # Python sees two identical string literals and points them to the same memory.
    a = "asdf"
    b = "asdf"
    print(f"2. Literal String Check ('asdf'): {a is b}")  
    # Output: True
    
    # 3. String Interning (Input - The "Run Time" Logic)
    # 'a' is created at compile time. 'b' is created when the code RUNS.
    # Python creates a NEW object for input, even if contents are identical.
    a = "manali"
    print("\n--- Interactive Check ---")
    print("Please enter 'manali' below for the test:")
    
    # Note: If you run this script, type 'manali' and hit enter
    # We simulate input here for demonstration:
    b = input() 
    # For this static code block, we simulate runtime creation:
    # b = "".join(['m', 'a', 'n', 'a', 'l', 'i']) 
    
    print(f"Content Check (a == b): {a == b}") # True (Values match)
    print(f"Memory Check  (a is b): {a is b}") # False (Different Objects!)


    print("\n=== PART 2: The 'in' Operator (Membership) ===\n")

    # 4. Dictionary Key vs. Value Trap
    data = {"key": "value"}

    # The Trap: 'in' only checks keys by default
    if "value" in data:
        print("Found value!")
    else:
        print("4. Trap: 'value' NOT found in data (It only checks keys!)")

    # The Fix: Explicitly ask for values
    if "value" in data.values():
        print("5. Fix: 'value' found in data.values() (Correct way)")


    print("\n=== PART 3: Integer Identity (The Cache) ===\n")
    
    # 6. Small Integers (Cached -5 to 256)
    x = 100
    y = 100
    print(f"6. Small Ints (100 is 100): {x is y}") # True

    # 7. Large Integers (Not Cached standardly)
    m = 257
    n = 257
    print(f"7. Large Ints (257 is 257): {m is n}") 
    # Likely False (though some modern IDEs/files might optimize this)

master_comparison_with_nuances()
```

## if - elif - else Statements

Python's conditional logic works similarly to C or Java (`else if` becomes `elif`), but with strict **short-circuit evaluation**.

**The Flow:**
1.  **`if`:** The entry point. Checked first.
2.  **`elif`:** Checked sequentially **only if** the previous `if` (and preceding `elif`s) were False.
    * *Critical Rule:* Python executes the block for the **first** True condition it finds and immediately **skips** the rest of the chain.
3.  **`else`:** The catch-all. Executed **only if** every single `if` and `elif` above it was False.

```python
x = 10

if x > 2:
    print("Greater than 2") # This is True! Prints and STOPS.
elif x > 5:
    print("Greater than 5") # This is NEVER checked.
else:
    print("Else block")
```

## Structural Pattern Matching

For structural pattern matching Python have the `match` statement, it looks like similar to switch statement in C and Java but it can do more than that

```python
def http_error(status):
    match status:
        case 400:
            return "Bad request"
        case 401 | 403 | 404:      #Can add multilple cases with or (|)
            return "Not found"
        case 418:
            return "I'm a teapot"
        case _:
            return "Something's wrong with the internet"

print(http_error(400)) #will go to matching 400 case block
print(http_error(403)) #will go to matching 403 case block
print(http_error(500)) #if nothing matches it will execute the wildcard (case _) and If no match exists, the behavior is a no-op and returns none
```
