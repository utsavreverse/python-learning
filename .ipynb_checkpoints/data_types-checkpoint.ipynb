# The Standard Type Hierarchy

## 1. The Fundamental Special Types

These are "singletons," meaning only one instance of them exists in memory during your program's life.

- **None** : Represents the absence of a value. It has no attributes and is the default return value for functions that don't return anything.

- **NotImplemented** : This is a special value used in numeric methods (like `__add__`). If you try to add A + B and A doesn't know how to handle B, it returns NotImplemented. This tells Python to try asking B to handle A instead.

- **Ellipsis** : Written literally as `....` It is mostly used in advanced slicing (like in NumPy arrays) or as a placeholder in code (e.g., `def my_func(): ...`).

```python
class Box:
    def __init__(self, value=None):
        # 1. Using None: specifically to represent "nothing here"
        if value is None:
            self.content = "Empty"
        else:
            self.content = value

    def __add__(self, other):
        # 2. Using NotImplemented: 
        # We only know how to add two Boxes together.
        # If 'other' is not a Box, we return NotImplemented so Python 
        # can try to solve it another way (or raise an error).
        if not isinstance(other, Box):
            return NotImplemented
        
        # If it IS a box, we combine them
        return Box(self.content + " & " + other.content)

    def teleport(self):
        # 3. Using Ellipsis:
        # We want to define this function structure, but we haven't
        # written the complex logic yet. '...' is valid code that does nothing.
        ... 

# --- Testing the types ---

# A. Using None
box_1 = Box("Toys")
box_2 = Box()        # value defaults to None
print(f"Box 2 contains: {box_2.content}")

# B. Using NotImplemented
# This works (Box + Box)
combined_box = box_1 + box_2 
print(f"Combined: {combined_box.content}")

# This fails gracefully (Box + Integer)
# Python sees NotImplemented and eventually raises a TypeError
try:
    result = box_1 + 42 
except TypeError:
    print("Error: You cannot add a number to a Box!")

# C. Using Ellipsis
# Calling the function with '...' inside runs fine, it just does nothing.
box_1.teleport()
print("Teleport function called (it did nothing).")
```

## 2. Numbers

Python treats numbers as objects. They are all immutable (you cannot change a number object; you can only create a new one).

- **Integers (int)** : These have unlimited precision. You can store a number as massive as your computer's memory allows.

- **Booleans (bool)** : Surprisingly, Booleans are a subtype of Integers. False behaves exactly like 0 and True behaves like 1.

- **Real (float)** : These are machine-level double-precision floating-point numbers (usually 64-bit).

- **Complex (complex)** : Numbers with a real and imaginary part (e.g., `3 + 4j`).

```python
#int data type
x = 5
print(type(x))

#in v2 it was long from v3 onwards it is int
x = 187687654564658970978909869576453
print(type(x))

#bool data type
x = True
print(type(x))

#float (real) data type
x = 1.34763
print(type(x))

#complex data type
x = 5 + 2j
print(type(x))
```

## 3. Sequences

Sequences are ordered collections of items indexed by non-negative integers. The documentation splits these into two crucial categories: Immutable and Mutable.
1. Immutable Sequences (Cannot be changed after creation)

- **Strings (str)** : A sequence of Unicode characters.
- **Tuples (tuple)** : A sequence of arbitrary Python objects.
- **Bytes (bytes)** : A sequence of 8-bit integers (0-255). Used heavily for binary data and file I/O.

2. Mutable Sequences (Can be changed in place)

- **Lists (list)** : The flexible arrays you use daily. They can hold any object and grow/shrink.
- **Byte Arrays (bytearray)** : The mutable version of bytes. You can modify specific bytes without creating a new object.

```python
"""Bytes and Bytearrays"""

# --- 1. BYTES (Immutable, like Strings) ---
b1 = b"Hello"                 # Literal syntax (note the 'b' prefix)
b2 = bytes(5)                 # Creates 5 null bytes: b'\x00\x00\x00\x00\x00'
b3 = bytes([65, 66, 67])      # From list of ints: b'ABC' (65='A')
# b1[0] = 72                  # ERROR: Bytes are immutable!

# --- 2. BYTEARRAY (Mutable, like Lists) ---
ba1 = bytearray(b"Hello")     # Create from existing bytes
ba2 = bytearray(5)            # Create 5 null bytes (mutable)
ba1[0] = 87                   # Works! 87 is ASCII for 'W', so ba1 is now b'Wello'

# --- 3. Encoding & Decoding ( The Bridge ) ---
s_text = "Café"               # String (Unicode)
b_data = s_text.encode('utf-8') # String -> Bytes (b'Caf\xc3\xa9')
s_back = b_data.decode('utf-8') # Bytes -> String ("Café")

# --- 4. Operations ---
b_cat = b1 + b"World"         # Concatenation
b_rep = b"Yo" * 3             # Repetition: b'YoYoYo'
val_int = b1[0]               # Accessing index returns an INT (0-255), not a char!
                              # b1[0] is 72, NOT b'H'

# --- 5. Slicing (Same as String) ---
b_slice = b1[1:4]             # b'ell' (Returns bytes object)
b_rev   = b1[::-1]            # b'olleH'

# --- 6. Common Methods (Most String methods work here too) ---
b_upper = b1.upper()          # b'HELLO'
b_split = b"A,B,C".split(b",")# [b'A', b'B', b'C']
b_find  = b1.find(b"l")       # Returns index of first 'l' (2)
b_hex   = b1.hex()            # "48656c6c6f" (Hexadecimal string representation)

# --- 7. Mutable Operations (Only for Bytearray) ---
ba1.append(33)                # Add '!' (ASCII 33) to the end
ba1.extend(b" World")         # Add multiple bytes
ba1.pop()                     # Remove last byte
ba1.reverse()                 # Reverse in-place

# --- PRINTING ---
print(f"Bytes: {b1} | First Byte (Int): {val_int}")
print(f"Encoded: {b_data} | Hex: {b_hex}")
print(f"Mutable Bytearray: {ba1}")
```

## 4. Set Types

Sets are unordered collections of unique and immutable objects. They are optimized for checking "is this item in the collection?" very quickly.

- **Sets (set)** : A mutable collection. You can add or remove items, but the items inside must be immutable (hashable).

- **Frozen Sets (frozenset)** : An immutable version of a set. Once created, you cannot change it. Because it is immutable, it can be used as a dictionary key (regular sets cannot).

```python
"""Set and Frozenset"""

set1 = set()                   # A new empty set
fset1 = frozenset()            # Empty frozenset
set1.add("cat")                # Add a single member
fset2 = frozenset([1, 2, 3])   # Frozenset with values in the list
#fset2.add(4)                  # Will raise an error since frozen sets are immutable
set1.update(["dog", "mouse"])  # Add several members, like list's extend
set1 |= set(["doe", "horse"])  # Add several members 2, like list's extend
if "cat" in set1:              # Membership test
  set1.remove("cat")
#set1.remove("elephant") - throws an error
set1.discard("elephant")       # No error thrown
print(set1)
for item in set1:              # Iteration AKA for each element
  print(item)
print("Item count:", len(set1))# Length AKA size AKA item count
#1stitem = set1[0]             # Error: no indexing for sets
isempty = len(set1) == 0       # Test for emptiness
set1 = {"cat", "dog"}          # Initialize set using braces; since Python 2.7
#set1 = {}                     # No way; this is a dict
set1 = set(["cat", "dog"])     # Initialize set from a list
set2 = set(["dog", "mouse"])
set3 = set1 & set2             # Intersection
set4 = set1 | set2             # Union
set5 = set1 - set3             # Set difference
set6 = set1 ^ set2             # Symmetric difference
issubset = set1 <= set2        # Subset test
issuperset = set1 >= set2      # Superset test
set7 = set1.copy()             # A shallow copy
set7.remove("cat")
print(set7.pop())              # Remove an arbitrary element
set8 = set1.copy()
set8.clear()                   # Clear AKA empty AKA erase
set9 = {x for x in range(10) if x % 2} # Set comprehension; since Python 2.7
print(set1, set2, set3, set4, set5, set6, set7, set8, set9, fset2, issubset, issuperset)
```

## 5. Mappings

Mappings represent collections of objects indexed by arbitrary keys (unlike sequences, which use integer indexes).

- **Dictionaries (dict)** : This is the only standard mapping type currently in Python.\
  **Keys** : Must be immutable (hashable) types like strings, numbers, or tuples.\
  **Values** : Can be anything.\
  **Ordering** : Since Python 3.7, dictionaries preserve insertion order (if you insert A then B, iterating over the dict gives you A then B).

```python
"""Dictionaries"""

dict1 = {} #Creates an empty dictionary
dict2 = dict() #Also creates an empty dictionary
dict2 = {"r": 34, "i": 56} #Initializing values
dict3 = dict(r=34, i=56) #Int
dict4 = dict([("r", 34), ("i",56)]) #Init from list of tuples
dict1["temprature"] = 32 #Assigning value to the key
if "temprature" in dict1: #membership check
    del dict1["temprature"]
equalbyvalue = dict2 == dict3
itemcount2 = len(dict2)        # Length AKA size AKA item count
isempty2 = len(dict2) == 0     # Emptiness test
for key in dict2:              # Iterate via keys
  print (key, dict2[key])        # Print key and the associated value
  dict2[key] += 10             # Modify-access to the key-value pair
for key in sorted(dict2):      # Iterate via keys in sorted order of the keys
  print (key, dict2[key])        # Print key and the associated value
for value in dict2.values():   # Iterate via values
  print (value)
for key, value in dict2.items(): # Iterate via pairs
  print (key, value)
dict5 = {} # {x: dict2[x] + 1 for x in dict2 } # Dictionary comprehension in Python 2.7 or later
dict6 = dict2.copy()             # A shallow copy
dict6.update({"i": 60, "j": 30}) # Add or overwrite; a bit like list's extend
dict7 = dict2.copy()
dict7.clear() # Clear AKA empty AKA erase
list1 = ['a','b']
dict8 = {(list1[0],list1[1]): 1, frozenset(list1): 2} #Keys can only be immutable data types like string, num, tuple and frozenset
#dict9 = {list1: 1, set(list1): 2} #This will raise an error because list and set are mutable entities
sixty = dict6.pop("i")         # Remove key i, returning its value
print (dict1, dict2, dict3, dict4, dict5, dict6, dict7, dict8, equalbyvalue, itemcount2, sixty)
```
