# The Standard Type Hierarchy

## 1. The Fundamental Special Types

These are "singletons," meaning only one instance of them exists in memory during your program's life.

- **None** : Represents the absence of a value. It has no attributes and is the default return value for functions that don't return anything.

- **NotImplemented** : This is a special value used in numeric methods (like `__add__`). If you try to add A + B and A doesn't know how to handle B, it returns NotImplemented. This tells Python to try asking B to handle A instead.

- **Ellipsis** : Written literally as `....` It is mostly used in advanced slicing (like in NumPy arrays) or as a placeholder in code (e.g., `def my_func(): ...`).

```python
class Box:
    def __init__(self, value=None):
        # 1. Using None: specifically to represent "nothing here"
        if value is None:
            self.content = "Empty"
        else:
            self.content = value

    def __add__(self, other):
        # 2. Using NotImplemented: 
        # We only know how to add two Boxes together.
        # If 'other' is not a Box, we return NotImplemented so Python 
        # can try to solve it another way (or raise an error).
        if not isinstance(other, Box):
            return NotImplemented
        
        # If it IS a box, we combine them
        return Box(self.content + " & " + other.content)

    def teleport(self):
        # 3. Using Ellipsis:
        # We want to define this function structure, but we haven't
        # written the complex logic yet. '...' is valid code that does nothing.
        ... 

# --- Testing the types ---

# A. Using None
box_1 = Box("Toys")
box_2 = Box()        # value defaults to None
print(f"Box 2 contains: {box_2.content}")

# B. Using NotImplemented
# This works (Box + Box)
combined_box = box_1 + box_2 
print(f"Combined: {combined_box.content}")

# This fails gracefully (Box + Integer)
# Python sees NotImplemented and eventually raises a TypeError
try:
    result = box_1 + 42 
except TypeError:
    print("Error: You cannot add a number to a Box!")

# C. Using Ellipsis
# Calling the function with '...' inside runs fine, it just does nothing.
box_1.teleport()
print("Teleport function called (it did nothing).")
```

## 2. Numbers

Python treats numbers as objects. They are all immutable (you cannot change a number object; you can only create a new one).

- **Integers (int)** : These have unlimited precision. You can store a number as massive as your computer's memory allows.

- **Booleans (bool)** : Surprisingly, Booleans are a subtype of Integers. False behaves exactly like 0 and True behaves like 1.

- **Real (float)** : These are machine-level double-precision floating-point numbers (usually 64-bit).

- **Complex (complex)** : Numbers with a real and imaginary part (e.g., `3 + 4j`).

```python
#int data type
x = 5
print(type(x))

#in v2 it was long from v3 onwards it is int
x = 187687654564658970978909869576453
print(type(x))

#bool data type
x = True
print(type(x))

#float (real) data type
x = 1.34763
print(type(x))

#complex data type
x = 5 + 2j
print(type(x))
```

## 3. Sequences

Sequences are ordered collections of items indexed by non-negative integers. The documentation splits these into two crucial categories: Immutable and Mutable.
1. Immutable Sequences (Cannot be changed after creation)

- **Strings (str)** : A sequence of Unicode characters.
- **Tuples (tuple)** : A sequence of arbitrary Python objects.
- **Bytes (bytes)** : A sequence of 8-bit integers (0-255). Used heavily for binary data and file I/O.

2. Mutable Sequences (Can be changed in place)

- **Lists (list)** : The flexible arrays you use daily. They can hold any object and grow/shrink.
- **Byte Arrays (bytearray)** : The mutable version of bytes. You can modify specific bytes without creating a new object.

```python
"""String"""

str1 = "Hello"                 #new string using double quotes
str2 = 'Hello'                 #single quotes do the same thing
str3 = 'Meow!\n\'Cat said..\'' # you can use \ to escape inbuilt characters \n to escape line and \t to escape tab
str4 = str1 + "World!"         #concatination
str5 = "Py" "thon"             #this will also work, but can't concatinate like this with variables
str6 = str1 + str(4)           #Explicit typecasting and concationation with a number
str7 = str1[2]                 #print character at second index
str7a = str1[-1]               #print last character
#str1[0] = "M"                 #No way! strings are immutable
for char in str1 : print(char) #Prints all the characters of str1
str8 = str1[1:]                #without first character
str9 = str1[:-1]               #without last character
str10 = str1[2:4]              #from index 2 to 3 (last index excluded)
str11 = str1[0::2]             #from starting with jump of two indexes
str12 = 3*"un"+"uim"           #Repetation and concatination
str13 = str1.lower()           #Converts to lowercase
str14 = str1.upper()           #Converts to uppercase
str15 = str1.capitalize()      #Capitalize the first letter
str16 = str1.lstrip()          #Remove leading blanckspaces
str17 = str4.rstrip("World!")  #remove this string from the right side
str18 = str1.strip()           #stripping from the both the sides
str19 = str3.replace(".","")   #replacing every . character to the blanck
list20 = str1.split("l")       #split the string from all the l's and add it to the list
if str1 == str2 : print("Eq")   #equality test
if "el" in str1 : print("In")  #Inclusion test
length = len(str1)             # Length
pos1 = str1.find('llo')        # Index of substring or -1
pos2 = str1.rfind('l')         # Index of substring, from the right
count = str1.count('l')        # Number of occurrences of a substring

print("---------------------------------")
print(str1,"\n",str2,"\n",str3,"\n",str4,"\n",str5,"\n",str6,"\n",str7,"\n",str7a,"\n",str8,"\n",str9,"\n",str10)
print("---------------------------------")
print(str11,"\n",str12,"\n",str13,"\n",str14,"\n",str15,"\n",str16,"\n",str17,"\n",str18,"\n",str19,"\n",list20)
print("---------------------------------")
print("Length:",length,"\n","Find llo at:",pos1,"\n","RFind l at:",pos2,"\n","Count of l:",count)
```

## 4. Set Types

Sets are unordered collections of unique and immutable objects. They are optimized for checking "is this item in the collection?" very quickly.

- **Sets (set)** : A mutable collection. You can add or remove items, but the items inside must be immutable (hashable).

- **Frozen Sets (frozenset)** : An immutable version of a set. Once created, you cannot change it. Because it is immutable, it can be used as a dictionary key (regular sets cannot).

## 5. Mappings

Mappings represent collections of objects indexed by arbitrary keys (unlike sequences, which use integer indexes).

- **Dictionaries (dict)** : This is the only standard mapping type currently in Python.\
  **Keys** : Must be immutable (hashable) types like strings, numbers, or tuples.\
  **Values** : Can be anything.\
  **Ordering** : Since Python 3.7, dictionaries preserve insertion order (if you insert A then B, iterating over the dict gives you A then B).
