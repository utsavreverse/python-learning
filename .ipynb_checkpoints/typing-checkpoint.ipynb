# Type Check in Python

## Data types
There are diffrent data types possible in various langauges, for convention (and most importantly to predict that what will happen after execution) langauges prefer the strongly type option, means there are some rules for using diffrent data types togather and we can't add them or concatenate them.(atleast directly or implicitly).

- **Strongly typed langauges** : C, Java, Python are strongly typed langauges\
for e.g. `2+'2'` will throw error in this langauges

- **Weekly typed langauges** : JavaScript is example of weekly typed langauge\
for e.g. `2+'2'` will give output `22`  (I can smell the JavaScript memes coming from miles away)

## Static and Dynamically typed langauges
Now if we talk about strongly typed langauges, there are also two more catagory exists in that.

1. **Static Typed** \
Static typed langauges are langauges that do type checking before runtime for e.g. langauges like C, C++ and Java

2. **Dynamically Typed** \
Dynamically typed langauges do type checking during runtime or execution for e.g. langauges like Python

Now we already seen that Python is dynamically typed langauge, so what does that mean? well that actually means that it interpret the code line by line so if type error not occurring while execution, it will simply runs the code.

```python
#This will run sucessfully
def foo(a):
    if (a>0):
        print(a+a)
    else:
        print(a+str(a))
foo(2) #This will run sucessfully
#foo(-1) #This will throw a type error
```

## Performance

Since statically typed langauges catches error befor runtime by type checking they don't need to check live during runtime and have laverage over the Dynamically typed langauges
