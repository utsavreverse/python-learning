```python
#5 Create a Python function to check if a given string is a palindrome

def is_palindrome(string_input):
    if string_input == string_input[::-1]:
        print("Given string is palindrome")
    else:
        print("Given string is not a palindrome")


string_input = input("Enter the string, note that comparision is a case sensitive: ")
is_palindrome(string_input)
```

```python
#6 Calculate the compound interest for a given principal amount, interest rate, and time period

while True:
    try:
        principal_amount = float(input("Enter the Principal Amount : "))
        interest_rate = float(input("Enter the Interest Rate : "))
        time_period = float(input("Enter the Time Period (in years) : "))
        break
    except ValueError:
        print("\nAll the values should be in numeric only\n")

compound_interest = (principal_amount * (1 + (interest_rate / 100)) ** time_period) - principal_amount
print("\n"+f"The compound interest for given data will be {compound_interest:.2f}")
```

```python
#7 Write a program that converts a given number of days into years, weeks, and days

"""The preassumtions of conditions are that year have 365 days, month have 30 days, week have 7 days"""

while True:
    try:
        days = int(input("Enter number of days : "))
        break
    except ValueError:
        print("\nOnly numerical values are accepted, please try again.\n")

if days >= 365:
    print(f"{days//365} years")
    print(f"{days//30} months")
    print(f"{days//7} weeks")
    print(f"{days} days")
elif days >= 30:
    print(f"{days//30} months")
    print(f"{days//7} weeks")
    print(f"{days} days")
elif days >= 7:
    print(f"{days//7} weeks")
    print(f"{days} days")
else:
    print(f"{days} days")
```

```python
#21 Create a program that takes a year as input and checks if it is a leap year or not

while True:
    try:
        year = input("Enter the year : ")
        if len(year) != 4:
            raise ValueError("Enter the valid year value")
        else:
            year = int(year)
            break
    except ValueError:
        print("\n Enter the valid year.\n")

if ((year % 400 == 0) or (year % 4 == 0)) and (year % 100 != 0):
    print(f"The year {year} is a leap year.")
else:
    print(f"The year {year} is not a leap year.")
```

```python
#24 Create a program that generates the Fibonacci sequence up to a given number of terms)

def fib(num):
    first_num, second_num = 0, 1
    while num > 0:
        print(first_num, end=' ')
        first_num, second_num = second_num, first_num+second_num
        num -= 1

while True:
    try:
        num = int(input("Enter the number of terms upto which you want a fibonacci sequence : "))
        if num <= 0:
            raise ValueError
        break
    except ValueError:
        print("\nThe valid number should be integer and grater than zero (i.e 1,2,3 etc)\n")

print(f"\nFibonacci sequance upto {num} terms:")
fib(num)
```
