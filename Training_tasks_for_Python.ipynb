## Implement stack

```python
# create naive script here

# only accepts the numeric input

input_stack = []

while True:
    print("""
------------ STACK -------------
1 . For pushing the values
2 . For popping values
3 . For current stack showing
4 . For exit""")
    choice = int(input("Please select your choice : "))

    match choice:
        case 1:
            try:
                val = int(input("Enter the number you want to push in the stack"))
                input_stack.append(val)
                print("Given value pushed sucessfully.")
            except ValueError:
                print("Enter the valid numeric value")

        case 2:
            try:
                popped_val = input_stack.pop()
                print(f"Value {popped_val} popped sucsessfully!")
            except IndexError:
                print("The stack is empty")

        case 3:
            for i in range((len(input_stack)-1), -1, -1):
                print(input_stack[i])
                print("-----")

        case 4:
            break

        case _:
            print("Please give valid input.")
```

```python
# create class based script here

class UnderflowError(Exception):
    pass

class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class Stack:
    def __init__(self):
        self.top = None

    def push(self, value):
        new_value = Node(value)
        new_value.next = self.top
        self.top = new_value

    def pop(self):
        try:
            if self.top is None:
                raise UnderflowError
            else:
                popped_value = self.top.value
                self.top = self.top.next
                return popped_value
        except UnderflowError:
            print("The stack is empty.")

    def show(self):
        if self.top is None:
            return "The stack is empty"

        status_print = self.top
        stack_showing = "Top -> "
        while status_print is not None:
            stack_showing += str(status_print.value) + " -> "
            status_print = status_print.next
        stack_showing += "Bottom"
        return stack_showing

input_stack = Stack()

while True:
    print("""
------------ STACK -------------
1 . For pushing the values
2 . For popping values
3 . For current stack showing
4 . For exit""")
    choice = int(input("Please select your choice : "))

    match choice:
        case 1:
            try:
                val = int(input("Enter the number you want to push in the stack"))
                input_stack.push(val)
                print("Given value pushed sucessfully in the stack.")
            except ValueError:
                print("Enter the valid numeric value")

        case 2:
            popped_val = input_stack.pop()
            if popped_val is not None:
                print(f"Value {popped_val} popped sucsessfully!")

        case 3:
            crr_stack = input_stack.show()
            print(crr_stack)

        case 4:
            break

        case _:
            print("Please give valid input.")
```

## W.A.P. that replicates a complete banking system. User should be able to perform below given tasks.

- __Open account__
- __login using account number and password__
- __logout__
- __add/withdraw money__
- __close account__
- __transfer amount__
- __keep track of transactions (all)__
- __maintain total amount of money that bank has__
- __take loan from bank at 7% interest rate for a time periode and pay EMIs__
- __Do this without an external database__
- __Use a JSON file as database__
- __Use postgresql as database__


```python
import random
import datetime
import json
import hashlib

class Account:
    __password = ""
    _account_number = 0

    def __init__(self):
        pass

    def __create_new_pass(self):
        while True:
            __passwd_input = input("Enter the password : ")
            __passwd_input_check = input("Re-enter your password : ")
            if __passwd_input == __passwd_input_check:
                password_hash = hashlib.sha512()
                password_hash.update(__passwd_input.encode('utf-8'))
                self.__password = password_hash.hexdigest()
                break
            else:
                print("Both password and re-entering should be the same, please try again.")

    def __create_new_accno(self):
        __assign_accno = random.randint(10**(14), (10**15) - 1)
        self._account_number = __assign_accno

    def __update_new_accounts(self):
        with open("account_details.json","a") as jw:
            new_acc = { self._account_number : self.__password}
            json.dump(new_acc, jw, indent=4)


    def final_msg(self):
        print("\nThank you for using our banking services :D\n")

    def create_account(self):
        choice = input("You are creating new account in the bank, do you wish to proceed ? [y/n]")
        str.lower(choice)
        if choice == "y" or choice == "yes":
            self.__create_new_accno()
            self.__create_new_pass()
            self.__update_new_accounts()

        elif choice == "n" or choice == "no":
            self.final_msg()
        
        else:
            print("You enterd the wrong choice please try again.")
            self.create_account()

acc1 = Account()
acc1.create_account()
```
