```python
# Defining some data sets for examples
import string

l1 = list(range(1, 11))
l2 = [i for i in string.ascii_lowercase if i<='j'] #list comprehension

t1 = tuple(range(1, 11))

d1 = dict(zip(l2, l1))
```

## For loop

For loop in python have capability to iterate over items in any sequance in order they appeard in the sequance.

```python
#Iterating over list
print("Items in list l1 : ")
for i in l1:
    print(i, end=" ")

#Iterating over tuple
print("\n\nItems in tuple t1 : ")
for i in t1:
    print(i, end=" ")

#Iterating over dict
print("\n\nKeys in dict d1 : ")
for i in d1: #d1.keys() will also give the same result
    print(i, end=" ")

print("\n\nValues in dict d1 : ")
for i in d1.values():
    print(i, end=" ")

print("\n\nKey - Value pair in dict d1")
for i, j in d1.items():
    print(f"Key : {i} - Value : {j}")
```

```python
#You can always use range function with slicing parameters in for loop

for i in range(10):
    print(i, end= " ")

for i in range(10,20):
    print(i, end= " ")

for i in range(20,30,2): #with two increments
    print(i, end= " ")

for i in range(40,30,-2): #with negative increament
    print(i, end= " ")
```

## While Loop

While loop in python will be executed untill condition for that while loop becomes false.

```python
count = 0

#Printing items of list using while
print("\n\nItems in list l1 using while loop: ")
while count < len(l1):
    print(l1[count], end=" ")
    count += 1

#Printing items of tuple using while
count = 0
print("\n\nItems in tuple t1 using while loop: ")
while count < len(t1):
    print(t1[count], end=" ")
    count += 1

#Printing items of dict using while
#Although it is possible to use while loop for dict, it is not recommended
```

## break, continue and pass in loops

- `break` statement will terminate the loop and transfer the control to the statement immediately following the loop. <br>
- `continue` statement will skip the current iteration and transfer the control to the beginning of the loop for the next iteration. <br>
- `pass` statement is a null operation; it is used when a statement is syntactically required but no action is needed.

```python
count = 0

def something():
    pass #Using pass statement as a placeholder for future code

while True:
    if count > 20: #Just a safety check to avoid infinite loop
        break #breaking the loop if something goes wrong
    else:
        if count % 2 == 0:
            print(f"The number {count} is even.")
            count += 1
        else:
            count += 1
            continue #skipping the rest of the loop for odd numbers
```

## else clause in loops

In Python, loops can have an optional `else` clause that executes when the loop completes normally (i.e., not terminated by a `break` or similar statements like `return` or `raise`). The `else` block will run after the loop finishes iterating over all items or when the while loop condition becomes false.

```python
for i in range(30):
    if i % 2 == 0:
        print(f"The number {i} is even.")
else:
    print("Even numbers from 0 to 29 have been printed.") #note that this else is associated with for loop and not if statement
```

## Breaking the nested loops

In Python there are many ways to break the nested loops. two common ways are using a flag variable or using exceptions.

```python
"""Using the flag variable to break nested loops"""

found = False
for i in range(1, 6):
    for j in range(1, 6):
        if i * j == 12:
            found = True
            break
    if found:
        break

print(f"Found a pair (i, j) such that i * j = 12: ({i}, {j})")


"""Using exceptions to break nested loops"""

class BreakNestedLoop(Exception):
    pass

try:
    for i in range(1, 6):
        for j in range(1, 6):
            if i * j == 12:
                raise BreakNestedLoop
except BreakNestedLoop:
    print(f"Found a pair (i, j) such that i * j = 12: ({i}, {j})")
```

```python
"""Using generator to wrap the loops (advanced)"""

def my_loops():
    
    while True:
    
        for num in range(100):
            if num > 10:
                yield False
            else:
                yield num


for item in my_loops():
    if item is not False:
        print(item, end=' ')
    else:
        break
```

## Recursion

Recursion is a programming technique where a function calls itself in order to solve a problem. A recursive function typically has two main components: a base case that stops the recursion and a recursive case that continues the recursion by calling the function again with modified arguments.

```python
"""Checking the armstrong number using recursion"""

length = 0
def is_armstrong(num):
    if num <= 0:
        return 0
    else:
        return ((num % 10) ** length + is_armstrong(num // 10))
    
number = int(input("Enter a number to check if it is an armstrong number: "))
length = len(str(number))
result = is_armstrong(number)
if result == number:
    print(f"{number} is an armstrong number.")
else:
    print(f"{number} is not an armstrong number.")
```

### Where to use which loop?

- Use a `for` loop when you know the number of iterations in advance or when iterating over a sequence (like a list, tuple, or string).
- Use a `while` loop when the number of iterations is not known beforehand and depends on a condition being true.
