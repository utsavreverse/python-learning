# Functions

- A function is a block of organized, reusable code that is used to perform a single, related action. Functions provide better modularity for your application and a high degree of code reusing.

- You can define functions to provide the required functionality. Here are important advantages of using functions:
  - __Code Reusability__: Once a function is defined, it can be used multiple times without rewriting the code.
  - __Modularity__: Functions allow you to break down complex problems into smaller, manageable parts.
  - __Readability__: Functions help in organizing code, making it easier to read and understand.
  - __Maintainability__: Changes can be made in one place (the function definition) rather than throughout the code.

**Function defination in python:**

```python
def function_name(parameters):
    """docstring"""
    
    statement(s)
    return [expression]
```

The keyword `def` introduces a function definition. It must be followed by the function name and the parenthesized list of formal parameters. The statements that form the body of the function start at the next line, and must be indented.

The first statement of the function body can optionally be a string literal; this string literal is the function’s documentation string, or docstring. (although not mandatory, it is a good practice to include it in your functions)

The execution of the function introduces a new **local symbol table** used for the local variables of the function. The local variables are those variables which are defined inside a function. The local variables exist only until the function is executing. When the function returns, the local symbol table is destroyed, and all the variables in it are forgotten.

Variable refreances search first in the local symbol table, then in the local symbol tables of enclosing functions, then in the global symbol table, and finally in the table of built-in names. This is known as the **LEGB Rule** (Local, Enclosing, Global, Built-in).

Global variables and variables of enclosing functions cannot be directly assigned a value within a function (unless, for global variables, named in a `global` statement, or, for variables of enclosing functions, named in a `nonlocal` statement), although they may be referenced.

A function definition associates the function name with the function object in the current symbol table. The interpreter recognizes the object pointed to by that name as a user-defined function. you can create mutiple objects of the same function by calling it multiple times.

```python
def greet(name):
    return "Hello, " + name + "!"

greet("Alice")
greet1 = greet("Bob")
greet2 = greet("Charlie")
```
In this example, the function `greet` is called three times with different arguments, creating three separate function call objects.

````python
def foo():
    pass

print(foo())
````

If you think from other langauge's perspective, above code should be a procedure and not a function because it is not returning any value, but in Python every function returns some value, so if return is not explicitly defined, it will returns the `None`, writing value `None` normally suppressed by the interpreter.

You can return any type of object (object can be string, number, list, tuple or anything that falls under Python's definition of an object) from a function, there is no restriction.

````python
def add_num(a, b):
    print(a+b)

def add_num_with_return(a, b):
    return a + b


a = float(input())
b = float(input())
add_num(a, b)

print(add_num_with_return(a, b))
````

### Default argument values

In Python, you can provide default values for function parameters. This allows you to call the function without providing all the arguments, and the default values will be used for any missing arguments.

````python
def ask_ok(prompt, retries=4, reminder='Please try again!'): #here retries and reminder are default arguments
    while True:
        reply = input(prompt)
        if reply in {'y', 'ye', 'yes'}:
            return True
        if reply in {'n', 'no', 'nop', 'nope'}:
            return False
        retries = retries - 1
        if retries < 0:
            raise ValueError('invalid user response')
        print(reminder)

ask_ok('Do you really want to quit?')
ask_ok('Do you really want to quit?', 2)
ask_ok('Do you really want to quit?', reminder='Come on, only yes or no!')
````

The default values are evaluated at the point of function definition in the defining scope, see example below:

````python
i = 5

def f(arg=i):
    print(arg) # although i is 6 at the time of function call, 
               # but default value is evaluated at the time of function definition, so it prints 5

i = 6
f()
````

But if you have mutable objects as default arguments, it can lead to unexpected behavior because the same object is used across multiple calls to the function.

````python
def f(a, L=[]):
    L.append(a)
    return L

print(f(1)) # prints [1]
print(f(2)) # prints [1, 2]
print(f(3)) # prints [1, 2, 3]
````

To overcome this issue, you can use `None` as the default value and then create a new object inside the function if needed.

````python
def f(a, L=None):
    if L is None:
        L = []
    L.append(a)
    return L
````

### Positional and Keyword Arguments

Python functions can be called using positional arguments, keyword arguments, or a combination of both.

**Positional arguments** : These are the most common type of arguments. The values are assigned to the parameters based on their position in the function call, so the order matters, these arguments are also the "required" arguments" i.e. these arguments must be provided when calling the function, otherwise it will raise a `TypeError`,these arguments are of the form `function_name(value1, value2, ...)`

**Keyword arguments** : These allow you to specify the values for specific parameters by name, regardless of their position in the function definition. This makes the function calls more readable and allows you to skip some arguments if they have default values, these arguments are of the form `function_name(kwarg1=value1, kwarg2=value2, ...)`

````python
# Here voltage is a positional argument, while state, action, and type are keyword arguments with default values.
def parrot(voltage, state='a stiff', action='voom', type='Norwegian Blue'):
    print("-- This parrot wouldn't", action, end=' ')
    print("if you put", voltage, "volts through it.")
    print("-- Lovely plumage, the", type)
    print("-- It's", state, "!")
````

````python
parrot(1000)                                          # 1 positional argument
parrot(voltage=1000)                                  # 1 keyword argument
parrot(voltage=1000000, action='VOOOOOM')             # 2 keyword arguments
parrot(action='VOOOOOM', voltage=1000000)             # 2 keyword arguments
parrot('a million', 'bereft of life', 'jump')         # 3 positional arguments
parrot('a thousand', state='pushing up the daisies')  # 1 positional, 1 keyword
````

````python
#These type of calling will leads to errors
parrot()                     # required argument missing
parrot(voltage=5.0, 'dead')  # non-keyword argument after a keyword argument
parrot(110, voltage=220)     # duplicate value for the same argument
parrot(actor='John Cleese')  # unknown keyword argument
````

You can also not pass values to the single parameter multiple times, or pass a value to a parameter that does not exist in the function definition, these will lead to errors.

````python
def function(a):
    pass

function(0, a=0)
````

There are two spacial types of arguments in Python functions:
- __*args__ : This allows a function to accept any number of positional arguments. The arguments are passed as a **tuple** to the function.
- __\*\*kwargs__ : This allows a function to accept any number of keyword arguments. The arguments are passed as a **dictionary** to the function.

It is important to note that when using `*args` and `**kwargs` in function definitions, the order matters. `*args` must come before `**kwargs` in the function signature, below is the correct way to define a function with both `*args` and `**kwargs`:

```python
def function_name(pos1, pos2, *args, kwarg1=value1, kwarg2=value2, **kwargs):
    # function body
```

````python
def cheeseshop(kind, *arguments, **keywords):
    print("-- Do you have any", kind, "?")
    print("-- I'm sorry, we're all out of", kind)
    for arg in arguments:
        print(arg)
    print("-" * 40)
    for kw in keywords:
        print(kw, ":", keywords[kw])
````

````python
cheeseshop("Limburger", "It's very runny, sir.",
           "It's really very, VERY runny, sir.",
           shopkeeper="Michael Palin",
           client="John Cleese",
           sketch="Cheese Shop Sketch")
````

### Spacial Parameters

Python also supports spacial parameters that can be used to enforce certain calling conventions in functions:
- __Positional-Only Parameters__ : These parameters can only be specified positionally and cannot be used as keyword arguments. They are defined by placing a `/` in the function signature.
- __Keyword-Only Parameters__ : These parameters can only be specified using keyword arguments and cannot be passed positionally. They are defined by placing a `*` in the function signature.

```text
def f(pos1, pos2, /, pos_or_kwd, *, kwd1, kwd2):
      -----------    ----------     ----------
        |             |                     |
        |     Positional or keyword      |
        |                                   -- Keyword only
        --- Positional only
```

````python
def standard_arg(arg): # standard argument
    print(arg)

def pos_only_arg(arg, /): # positional-only argument
    print(arg)

def kwd_only_arg(*, arg): # keyword-only argument
    print(arg)

def combined_example(pos_only, /, standard, *, kwd_only): # combined example
    print(pos_only, standard, kwd_only)
````

````python
standard_arg(2)

standard_arg(arg=2)
````

````python
pos_only_arg(1)

pos_only_arg(arg=1)
````

````python
kwd_only_arg(arg=3)

kwd_only_arg(3)
````

````python
combined_example(1, 2, kwd_only=3) # correct
combined_example(1, standard=2, kwd_only=3) # correct

combined_example(1, 2, 3) # passing kwd_only as positional argument - incorrect
combined_example(pos_only=1, standard=2, kwd_only=3) # passing pos_only as keyword argument - incorrect
````

Now consider the below example:

````python
def foo(name, **kwds):
    return 'name' in kwds
````

````python
foo(1, **{'name': 2}) # this will raise TypeError because 'name' is passed as a positional argument
````

````python
def foo(name, /, **kwds):
    return 'name' in kwds

foo(1, **{'name': 2}) # this will work fine because 'name' is now a keyword-only argument
````

**Where to use when?**

- Use positional-only if you want the name of the parameters to not be available to the user. This is useful when parameter names have no real meaning, if you want to enforce the order of the arguments when the function is called or if you need to take some positional parameters and arbitrary keywords.

- Use keyword-only when names have meaning and the function definition is more understandable by being explicit with names or you want to prevent users relying on the position of the argument being passed.

- For an API, use positional-only to prevent breaking API changes if the parameter’s name is modified in the future.

### Documentation Strings

A documentation string (or docstring) is a special type of comment that is used to document a specific segment of code. In Python, docstrings are written using triple quotes (`"""` or `'''`) and are placed immediately after the function, method, class, or module definition.

You can access a function's docstring using the `__doc__` attribute or the built-in `help()` function. This is useful for understanding what a function does, its parameters, and its return values without having to look at the source code.

````python
def my_function():
    """Do nothing, but document it.

    No, really, it doesn't do anything:

        >>> my_function()
        >>>
    """
    pass

print(my_function.__doc__) # prints the docstring of my_function,
                           # help(my_function) can also be used to get more detailed help
````

### Lambda Expressions

A lambda expression is a small anonymous function that can take any number of arguments but can only have one expression. It is often used for short, throwaway functions that are not going to be reused elsewhere in the code.

````python
from functools import reduce

print("=== 1. The Basic Syntax (Assigning to variable) ===")
# Syntax: lambda arguments: expression
add = lambda x, y: x + y
print(f"3 + 5 = {add(3, 5)}")


print("\n=== 2. Using with Built-in Functional Tools (Map, Filter, Reduce) ===")
numbers = [1, 2, 3, 4, 5, 6]

# map(): Apply logic to every item
squared = list(map(lambda x: x**2, numbers))
print(f"Squared: {squared}")

# filter(): Keep items where logic is True
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(f"Evens:   {evens}")

# reduce(): Combine all items into one (e.g., sum or product)
total = reduce(lambda x, y: x + y, numbers)
print(f"Sum:     {total}")


print("\n=== 3. Sorting & Key Functions (The Most Common Use Case) ===")
students = [
    {'name': 'Alice', 'score': 90, 'age': 20},
    {'name': 'Bob', 'score': 75, 'age': 22},
    {'name': 'Charlie', 'score': 90, 'age': 19}
]

# Sort by Score (Ascending)
by_score = sorted(students, key=lambda s: s['score'])
print(f"Sorted by score: {[s['name'] for s in by_score]}")

# Sort by Score (Descending), then by Age (Ascending) for ties
# We use a tuple (negated_score, age) to sort
complex_sort = sorted(students, key=lambda s: (-s['score'], s['age']))
print(f"Complex sort:    {[s['name'] for s in complex_sort]}")

# Max/Min with key
oldest = max(students, key=lambda s: s['age'])
print(f"Oldest student:  {oldest['name']}")


print("\n=== 4. Immediate Invocation (IIFE) ===")
# Running a lambda immediately without assigning it (Rare but valid)
result = (lambda x: x * 2)(10)
print(f"Immediate call:  {result}")


print("\n=== 5. Returning a Function (Closures) ===")
def power_generator(n):
    # Returns a lambda that locks in 'n'
    return lambda x: x ** n

square_func = power_generator(2) # n=2
cube_func = power_generator(3)   # n=3

print(f"5 squared is: {square_func(5)}")
print(f"5 cubed is:   {cube_func(5)}")
````
