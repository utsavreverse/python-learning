Spacing is known as indentation in Python it is leading whitespaces using `space` or `tab` before code, it used to define scope in Python, in other langauges they use diffrent methods like in Java they use curly brackets `{}` for scope defination.

Note : Although we can use `space` or `tab` both for indentation but code formatting guideline of python [PEP8](https://peps.python.org/pep-0008/#indentation) recommands to use 4 spaces as a indentation

```python
for i in range(10):  # the first line 0 space
    if i==0:  # first level indent 4 space
        print("I am first")  # second level indent 8 space
    elif i==9:  # first level indent 4 space
        print("I am last")  # second level indent 8 space
    else:  # first level indent 4 space
        print("Where are we?")  # second level indent 8 space

# end of line
```
