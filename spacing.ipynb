Spacing is known as indentation in Python it is leading whitespaces using `space` or `tab` before code, it used to define scope in Python, in other langauges they use diffrent methods like in Java they use curly brackets `{}` for scope defination.

Note : Although we can use `space` or `tab` both for indentation but code formatting guideline of python [PEP8](https://peps.python.org/pep-0008/#indentation) recommands to use 4 spaces as a indentation

```python
for i in range(10):  # the first line 0 space
    if i==0:  # first level indent 4 space
        print("I am first")  # second level indent 8 space
    elif i==9:  # first level indent 4 space
        print("I am last")  # second level indent 8 space
    else:  # first level indent 4 space
        print("Where are we?")  # second level indent 8 space

# end of line
```

### [Indentation Backend](https://docs.python.org/3/reference/lexical_analysis.html#indentation)
(after 27 Jan 2026 review)

Before the first line of Python code is read, the parser initializes a **stack** and pushes `0` onto it. This `0` represents the base indentation level and is **never popped off**. This stack tracks the indentation level for every logical line.

When the parser encounters a new logical line, it compares the current indentation level to the top value of the stack:

1.  **Indentation > Top of Stack (Opening a Block)** <br>
    * The new indentation level is **pushed** onto the stack.
    * One `INDENT` token is generated.<br><br>

2.  **Indentation < Top of Stack (Closing a Block)** <br>
    * The parser **pops** values off the stack until it finds a number that matches the current indentation level.
    * For every value popped, a `DEDENT` token is generated.
    * *(Note: If no match is found, an `IndentationError` occurs).*<br><br>

3.  **Indentation == Top of Stack (Continuing a Block)**<br>
    * Nothing happens to the stack.
    * The parser simply continues to the next logical line.<br><br>

```python
# Example

def my_func():        # Line 1 (Indent: 0)
    if True:          # Line 2 (Indent: 4)
        print("Hi")   # Line 3 (Indent: 8)
    print("Done")     # Line 4 (Indent: 4)
print("End")          # Line 5 (Indent: 0)
```

**This will be the process in the stack for the above code**

| Line | Code | Indentation | Comparison | Action | Stack State (After) | Token Generated |
| :--- | :--- | :--- | :--- | :--- | :--- | :--- |
| **1** | `def ...` | **0** | `0 == Top(0)` | No Change | `[0]` | *(None)* |
| **2** | `if ...` | **4** | `4 > Top(0)` | **Push 4** | `[0, 4]` | `INDENT` |
| **3** | `print("Hi")` | **8** | `8 > Top(4)` | **Push 8** | `[0, 4, 8]` | `INDENT` |
| **4** | `print("Done")`| **4** | `4 < Top(8)` | **Pop 8** | `[0, 4]` | `DEDENT` |
| **5** | `print("End")` | **0** | `0 < Top(4)` | **Pop 4** | `[0]` | `DEDENT` |

### Few Notes on Indentation

* **Editor Consistency:**<br>
Different text editors may display tabs with different visual widths (e.g., some show 2 spaces, others 8). Therefore, **never mix tabs and spaces**. You should strictly use the same type of indentation throughout your code. Mixing them often leads to a `TabError` in Python 3. Unless explicitly required otherwise, always follow the **PEP 8 standard** (4 spaces). <br><br>

* **Dynamic Tab Expansion:**<br>
Python converts tabs to spaces based on "Tab Stops" at every 8th column. The visual width of a tab depends on where it is placed:
    * `Tab` + `Space`: The `Tab` fills 8 spaces (0 to 8), plus 1 `Space` = **9 total indentation**.
    * `Space` + `Space` + `Tab`: The 2 spaces take columns 0-1. The `Tab` fills the remaining 6 spots to reach column 8. **8 total indentation**.
    * **Note:** In Python 3, mixing tabs and spaces in a way that depends on this calculation will raise a `TabError`. <br><br>

* **Stack Flexibility:**<br>The Python parser does not enforce a specific number of spaces (like 4). As long as the new indentation level is **greater** than the current top of the stack, Python accepts it as a valid block.
    * *Example:* If the current level is 0, even **1 space** is sufficient to create an indent.
    * *Best Practice:* While valid, this is discouraged. Always stick to the PEP 8 standard (4 spaces) for readability and maintenance.
