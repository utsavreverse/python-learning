# Iterators and Iterables

### Iterables
An **iterable** is any Python object capable of returning its members one at a time, allowing it to be iterated over in a loop. Examples of iterables include lists, tuples, strings, and dictionaries.

### Backend processing

- In python in order for an object to be __iterable__ it must implement \_\_iter\_\_ method.
- \_\_iter\_\_ is expexted to return an __Iterator__ object
- Thus, An __iterable__ object is an object that implements \_\_iter\_\_
- Which brings us over to our next question what is an __Iterator__?

### Iterators
- An __iterator__ is an object that implements \_\_next\_\_ method.
- Which is expected to return the __next element of the iterable object__ that returned it.
- Raise a __StopIteration exception__ when no more elements are available.

```python
l1 = list(range(1,11))
print(dir(l1)) #this will show all the methods implemented by the iterator object, look carefully for __iter__

it_1 = iter(l1)  # l1.__iter__() is called here 
print(dir(it_1)) #this will show all the methods implemented by the iterator object, look carefully for __next__ as well as __iter__

print(next(it_1)) # it_1.__next__() is called here
print(it_1.__next__()) # this is equivalent to the above line
```

```python
it_1 = iter(l1)  # l1.__iter__() is called here
while True:
    try:
        element = next(it_1)  # it_1.__next__() is called here
        print(element)
    except StopIteration:
        break
```

```python
# Look above code carefully, that's how iteration of iterables done when you use for loop as code like below
for element in l1:
    print(element)
```

### How to make custom iterable object?

```python
"""Method 1: Implementing Iterable by defining __iter__ method which returns an Iterator object"""

class Team:

    def __init__(self):
        self._junior_players = list()
    
    def add_junior_player(self, player_name):
        self._junior_players.append(player_name)

    def __iter__(self):
        return TeamIterator(self._junior_players)
    
class TeamIterator:
    
    def __init__(self, junior_players):
        self._junior_players = junior_players # list of junior players
        self._index = 0 # to keep track of current index during iteration

    def __iter__(self):
        return self

    def __next__(self):
        if self._index < len(self._junior_players):
            player = self._junior_players[self._index]
            self._index += 1
            return player
        else:
            raise StopIteration
```

```python
team = Team()
team.add_junior_player("Alice")
team.add_junior_player("Bob")
team.add_junior_player("Charlie")

for player in team:
    print(player)

print("Alice" in team._junior_players)  # Direct access to the underlying list
```

```python
"""Method 2 (simpler): Implementing Iterable by defining __iter__ method which yields elements one by one"""

class Team:
    
    def __init__(self):
        self._junior_players = list()
    
    def add_junior_player(self, player_name):
        self._junior_players.append(player_name)

    def __iter__(self):
        return iter(self._junior_players)
    
    def __next__(self):
        return next(self)
```

```python
team = Team()
team.add_junior_player("Alice")
team.add_junior_player("Bob")
team.add_junior_player("Charlie")

for player in team:
    print(player)
```
