# Classes and Object Oriented Programming (OOP) in Python

### Classes and Objects
- A class is a blueprint for creating objects. It defines a set of attributes and methods that the objects created from the class will have.
- An object is an instance of a class. It is created using the class and can have its own unique values for the attributes defined in the class.

````python
#classes in python defined by 'class' keyword
class Example: #also it is standard to use capitalized names for classes, it is a convention in python
    pass

#creating an object of the class
obj = Example() #object creation
print(obj) # object is an instance of the class, it has its own memory location,
           # if anything is not returned, it will just print the memory location of the object.
````

**constructors**: A constructor is a special method in a class that is automatically called when an object of the class is created. It is used to initialize the attributes of the object. In Python, the constructor method is defined using the `__init__` method, so when the new instance of the class is created, the `__init__` method is automatically called to initialize the attributes of the object. The `__init__` method can take parameters to set the initial values of the attributes when the object is created.

````python
class Person:
    def __init__(self, name, age): #constructor method to initialize the attributes of the object
        self.name = name #self is a reference to the current instance of the class, it is used
                         #to access the attributes and methods of the class
        self.age = age


person1 = Person("Alice", 30) #creating an object of the class Person and passing the values for the attributes
print(person1.name) #accessing the name attribute of the person1 object
print(person1.age) #accessing the age attribute of the person1 object
````

Even without defining attributes in the constructor, we can create attributes for any instance later on. This is because Python allows us to add attributes to an instance dynamically, even after the object has been created. This means that we can create attributes for an instance at any point in time, and they will be specific to that instance only. However, it is generally recommended to define attributes in the constructor for better code organization and readability. However it is recommended to define attributes in the constructor for better code organization and readability, as it provides a clear structure for the class and makes it easier to understand the attributes that belong to each instance of the class.

````python
class ComplexNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag
````

````python
complex1 = ComplexNumber(2, 3) #creating an object of the class ComplexNumber and passing the values for the attributes
print(complex1.real) #accessing the real attribute of the complex1 object
print(complex1.imag) #accessing the imag attribute of the complex1 object
````

````python
complex1.some_attribute = "This is a new attribute" #creating a new attribute for the complex1 object,
                                                    #that is not defined in the constructor
print(complex1.some_attribute) #accessing the new attribute of the complex1 object
````

### Class and instance variables

- **Class variables**: Class variables are shared among all instances of a class. They are defined within the class but outside of any instance methods. Class variables are accessed using the class name or through any instance of the class. They are typically used to store information that is common to all instances of the class.

- **Instance variables**: Instance variables are unique to each instance of a class. They are defined within the constructor method (`__init__`) and are accessed using the `self` keyword. Each instance of the class can have different values for its instance variables, allowing for individual customization of each object created from the class.

````python
class Vehicle:
    type = "Car" #class variable

    def __init__(self, company, model):
        self.company = company #instance variable
        self.model = model #instance variable
````

````python
car1 = Vehicle("Toyota", "Camry") #creating an object of the class Vehicle and passing the values for the attributes
print(car1.type) #accessing the class variable type using the car1 object
print(car1.company) #accessing the instance variable company of the car1 object
print(car1.model) #accessing the instance variable model of the car1 object
````

````python
car2 = Vehicle("Honda", "Civic") #creating another object of the class Vehicle and passing different values for the attributes
print(car2.type) #accessing the class variable type using the car2 object
print(car2.company) #accessing the instance variable company of the car2 object
print(car2.model) #accessing the instance variable model of the car2 object
````

````python
#you can also access the class variable using the class name
print(Vehicle.type) # accessing the class variable type using the class name
                    # it is useful when you want to access the class variable without creating an object of the class


Vehicle.type = "Truck" #modifying the class variable type using the class name
print(car1.type) #accessing the modified class variable type using the car1 object
print(car2.type) #accessing the modified class variable type using the car2 object
````

If you change the class attribute using the class name, it will change for all instances of the class. However, if you change the class attribute using an instance, it will only change for that instance and not for the others. This is because when you access a class attribute through an instance, Python first looks for an instance attribute with the same name. If it finds one, it uses that instead of the class attribute. If it doesn't find one, it uses the class attribute. Therefore, changing the class attribute through an instance creates a new instance attribute that shadows the class attribute for that specific instance.

````python
class MyClass:
    a = None
    b = None
````

````python
obj1 = MyClass()
obj1.a = "ghi" #creating an instance variable a for the obj1 object and assigning a value to it
obj1.b = "jkl" #creating an instance variable b for the obj1 object and assigning a value to it
print(obj1.a) #accessing the class variable a using the obj1 object
print(obj1.b) #accessing the class variable b using the obj1 object
````

````python
MyClass.a = "abc" #assigning value to the class variable a using the class name
MyClass.b = "def" #assigning value to the class variable b using the class
````

````python
""" this will not change the values of a and b for the obj1 object 
because they are instance variables that were created for the obj1 object 
and they shadow the class variables a and b for that specific instance. """

print(obj1.a) #accessing the class variable a using the obj1 object after modifying it using the class name
print(obj1.b) #accessing the class variable b using the obj1 object after modifying
````

````python
obj2 = MyClass() #creating another object of the class MyClass
print(obj2.a) #accessing the class variable a using the obj2 object
              #it will print the modified value of a because obj2 does not have an instance variable a
              #so it will use the class variable a

print(obj2.b) #accessing the class variable b using the obj2 object
````

If you equate any function outside the class to an attribute of the class, it will become a method of the class. This is because in Python, functions are first-class objects, and when you assign a function to an attribute of a class, it becomes a method that can be called on instances of the class. The function will have access to the instance (through the `self` parameter) and can operate on the instance's attributes, making it a method of the class.


````python
# Function defined outside the class
def f1(self, x, y):
    return min(x, x+y)

class C:
    f = f1 #it will not return the value
           #rather it is whole function that is assigned to the attribute f of the class C
           # so it will become a method of the class C

    def g(self):
        return 'hello world'

    h = g  #similarly we can do with the methods inside of the class.
````

````python
c1 = C() #creating an object of the class C
print(c1.f(5, -1)) #calling the method f of the class C
print(c1.g()) #calling the method g of the class C
print(c1.h()) #calling the method h of the class C, it will return the
              #same output as the method g.
````

Methods can also call other methods of the same class. This is because when a method is called on an instance of a class, it has access to all the attributes and methods of that instance, including other methods defined in the class. This allows for code reuse and modularity within the class, as one method can call another method to perform a specific task without having to duplicate code, the syntax for calling another method within a method is simply to use `self.method_name()` where `method_name` is the name of the method you want to call.

````python
class Bag:
    def __init__(self):
        self.data = []

    def add(self, x):
        self.data.append(x)

    def addtwice(self, x):
        self.add(x)
        self.add(x)
````

### Inheritance

Inheritance is a fundamental concept in object-oriented programming (OOP) that allows a class (called the child class or subclass) to inherit attributes and methods from another class (called the parent class or superclass). This promotes code reusability and establishes a hierarchical relationship between classes.

````python
"""Example 1"""

class Car:
    def __init__(self, company, model):
        self.company = company
        self.model = model

class ElectricCar(Car):
    def __init__(self, company, model, battery_capacity):
        super().__init__(company, model)  # Call the constructor of the parent class
        self.battery_capacity = battery_capacity  # New attribute for ElectricCar

    def display_info(self):
        print(f"Company: {self.company}, Model: {self.model}, Battery Capacity: {self.battery_capacity} kWh")
````

````python
car1 = ElectricCar("Tesla", "Model S", 100)
car1.display_info()  # Output: Company: Tesla, Model: Model S, Battery Capacity: 100 kWh
````

**`isinstance()`** is a built-in function in Python that checks if an object is an instance of a specified class or a subclass thereof. It takes two arguments: the object to be checked and the class (or a tuple of classes) against which the object is being tested. The function returns `True` if the object is an instance of the specified class or any of its subclasses, and `False` otherwise.

**`issubclass()`** is a built-in function in Python that checks if a class is a subclass of another class. It takes two arguments: the first is the class to be checked, and the second is the class (or a tuple of classes) against which the first class is being tested. The function returns `True` if the first class is a subclass of the second class or any of its subclasses, and `False` otherwise.

````python
print(isinstance(car1, ElectricCar))  # Output: True
print(isinstance(car1, Car))          # Output: True
print(issubclass(ElectricCar, Car))   # Output: True
print(issubclass(Car, ElectricCar))   # Output: False
````

````python
"""Intresting take on isinstance"""

#to check whether the given input is number or not
def is_number(n):
    return isinstance(n, (int, float, complex)) and not isinstance(n, bool)

print(is_number(10))        # True
print(is_number(10.5))      # True
print(is_number(3 + 4j))    # True
print(is_number(True))      # False
print(is_number("Hello"))   # False
````

````python
"""Example 2"""

class A:
    def method_a(self):
        print(f"memory addr of obj in A: {id(self)}")
        print("method_a from A")

    def method_b(self):
        self.method_a()

class B(A):
    def method_a(self):
        print(f"memory addr of obj in B: {id(self)}")
        print("method_a from B")
#         A.method_a(self)
        super().method_a()
````

````python
b = B()
b.method_b()
print("Note that both the IDs are same, because the self in both the methods refer to the same object b of class B.")
````

**Multiple Inheritance**: Python supports multiple inheritance, which means that a class can inherit from more than one parent class. This allows for greater flexibility in designing classes and can help to avoid code duplication. However, it can also lead to complexity and ambiguity if not used carefully, especially when multiple parent classes have methods or attributes with the same name.

Python uses the C3 linearization algorithm (also known as the C3 superclass linearization) to determine the method resolution order (MRO) in cases of multiple inheritance, ensuring a consistent and predictable order for method lookups, which means that Python will search for methods and attributes in a specific order when multiple parent classes are involved, it will first look in the child class, then in the parent classes in the order they are listed in the class definition, and finally in the built-in `object` class.

````python
class Animal:
    def __init__(self, age, gender):
        self.age = age
        self.gender = gender

class Cat:
    def __init__(self, color):
        self.color = color

class StrayCat(Animal, Cat):
    def __init__(self, age, gender, color, name):
        Animal.__init__(self, age, gender)
        Cat.__init__(self, color)
        self.name = name
````

### Encapsulation

Encapsulation is one of the fundamental principles of object-oriented programming (OOP) that refers to the bundling of data (attributes) and methods (functions) that operate on that data within a single unit, typically a class. It restricts direct access to some of an object's components, which helps to prevent unintended interference and misuse of the data. Encapsulation is achieved through access modifiers that define the visibility of class members (attributes and methods). In Python, there are three levels of access control:

- **Public**: Members declared as public are accessible from anywhere, both inside and outside the class. In Python, all members are public by default.
- **Protected**: Members declared as protected are intended to be accessible only within the class and its subclasses. In Python, a single underscore prefix (e.g., `_member`) is used to indicate that a member is protected, but this is just a convention and does not enforce access restrictions.
- **Private**: Members declared as private are intended to be accessible only within the class itself. In Python, a double underscore prefix (e.g., `__member`) is used to indicate that a member is private.

```python
class MyClass:
    def __init__(self):
        self.public_attr = "I am public"
        self._protected_attr = "I am protected"
        self.__private_attr = "I am private"

    def public_method(self):
        return "This is a public method"

    def _protected_method(self):
        return "This is a protected method"

    def __private_method(self):
        return "This is a private method"
```

````python
class Car:
    __price = {1:890000, 2: 750000, 3: 600000}

    def __init__(self, id, company, model):
        self.id = id
        self.company = company
        self.model = model

    
    def __get_quote(self):
        return Car.__price.get(self.id, "Price not available")
    
    """Although direct access of the private or protected members (including methods and attributes)
    is not possible from outside the class, we can create public methods that internally access these members.
    This way, we can control how the private or protected members are accessed and provide a public interface
    for them."""
    
    def display_info(self):
        price = self.__get_quote()
        print(f"Company: {self.company}, Model: {self.model}, Price: {price}")
````

````python
car1 = Car(1, "BMW", "X5")
car1.display_info()  # Output: Company: BMW, Model: X5, Price: 890000
````

### Polymorphism

Polymorphism is a core concept in object-oriented programming (OOP) that allows objects of different classes to be treated as objects of a common superclass. It enables a single interface to represent different underlying data types or classes, allowing for flexibility and extensibility in code design. Polymorphism can be achieved through method overriding or method overloading.

- **Method Overriding**: This occurs when a subclass provides a specific implementation of a method that is already defined in its superclass. The method in the subclass has the same name, return type, and parameters as the method in the superclass. When the method is called on an instance of the subclass, the overridden method in the subclass is executed instead of the one in the superclass.

- **Method Overloading**: This is a feature that allows a class to have multiple methods with the same name but different parameters (either in number, type, or both). In Python, direct method overloading is not possible (unlike Java or C++), but you can achieve similar functionality using default arguments or variable-length arguments.

````python
class Shape:
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14 * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height
````

````python
shape1 = Circle(5)
shape2 = Rectangle(4, 6)
print(f"Area of the circle: {shape1.area()}") # Output: Area of the circle: 78.5
print(f"Area of the rectangle: {shape2.area()}") # Output: Area of the rectangle: 24
````

### Abstraction

Abstraction is a fundamental principle of object-oriented programming (OOP) that focuses on hiding the complex implementation details of a system and exposing only the necessary and relevant features to the user. It allows developers to create a simplified model of a complex system, making it easier to understand and work with.

In Python, abstraction can be achieved using abstract classes and interfaces. An abstract class is a class that cannot be instantiated and is meant to be subclassed. It can contain abstract methods, which are methods that are declared but not implemented in the abstract class. Subclasses of the abstract class are required to provide implementations for the abstract methods.

````python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height
````

````python
# 1. USAGE: You can create Circle and Rectangle objects and call their area methods.
shapes = [Circle(5), Rectangle(4, 6)]
for shape in shapes:
    print(f"Area: {shape.area():.2f}")

# 2. SAFETY CHECK: You strictly CANNOT create a generic Shape
# s = Shape() 
# TypeError: Can't instantiate abstract class Shape with abstract method area

# 3. ENFORCEMENT: If you make a Triangle but forget 'area', Python stops you immediately.
""" class Triangle(Shape):
    pass
t = Triangle()  """
# TypeError: Can't instantiate abstract class Triangle with abstract method area
````

### @classmethod decorator

A `@classmethod` is a method that belongs to the class rather than an instance of the class. It takes the class itself as the first argument, conventionally named `cls`, instead of the instance (which is conventionally named `self`). This allows the method to access and modify class-level attributes and methods. Class methods are defined using the `@classmethod` decorator above the method definition. They can be called on the class itself or on instances of the class, but they will always receive the class as the first argument.

````python
class Employee:

    num_of_emps = 0          # Always same for all the instances of class
    raise_amount = 1.04      # May differ from employee to employee

    def __init__(self, first_name, last_name, salary):
        self.first_name = first_name
        self.last_name = last_name
        self.salary = salary
        self.email = first_name + '.' + last_name + "@reversebits.com"

        Employee.num_of_emps +=  1


    def full_name(self):
        return f"{self.first_name} {self.last_name}"

    def apply_raise(self):
        print(f"Raising the salary with amount: {self.raise_amount}")
        self.salary = int(self.salary*self.raise_amount)

    @classmethod
    def set_raise_amount(cls, amount):
        cls.raise_amount = amount

    @classmethod
    def from_string(cls, string):
        first_name, last_name, salary = string.split('-')
        return cls(first_name, last_name, int(salary))
````

````python
emp_1 = Employee("John", "silvesa", "10000000")
emp_2 = Employee("Corey", "Schafer", "2000000")
print(f"emp_1.raise_amount = {emp_1.raise_amount}")
print(f"emp_2.raise_amount = {emp_2.raise_amount}")
````

````python
"""Notice how it changed for the both the instances"""

Employee.set_raise_amount(1.05)
print(f"emp_1.raise_amount = {emp_1.raise_amount}")
print(f"emp_2.raise_amount = {emp_2.raise_amount}")
````

### @staticmethod decorator

A `@staticmethod` is a method that does not take the instance (`self`) or the class (`cls`) as its first argument. It is a method that belongs to the class but does not have access to the instance or class attributes. Static methods are defined using the `@staticmethod` decorator above the method definition. They can be called on the class itself or on instances of the class, but they do not receive any implicit first argument. Static methods are typically used for utility functions that perform a task in isolation and do not require access to instance or class data.

Since it does not take `self` or `cls` as a parameter, it cannot access or modify instance or class attributes. It is essentially a regular function that happens to be defined within the class's namespace, and it is used when you want to perform a function that is related to the class but does not need to access any class or instance-specific data.

````python
class Employee:

    num_of_emps = 0          # Always same for all the instances of class
    raise_amount = 1.04      # May differ from employee to employee

    def __init__(self, first_name, last_name, salary):
        self.first_name = first_name
        self.last_name = last_name
        self.salary = salary
        self.email = first_name + '.' + last_name + "@reversebits.com"

        Employee.num_of_emps +=  1


    def full_name(self):
        return f"{self.first_name} {self.last_name}"

    def apply_raise(self):
        print(f"Raising the salary with amount: {self.raise_amount}")
        self.salary = int(self.salary*self.raise_amount)

    @classmethod
    def set_raise_amount(cls, amount):
        cls.raise_amount = amount

    @classmethod
    def from_string(cls, string):
        first_name, last_name, salary = string.split('-')
        return cls(first_name, last_name, int(salary))

    @staticmethod
    def is_weekend(day):
        return day.weekday() == 5 or day.weekday() == 6
````

### Dunder or Magic Methods

Dunder methods, also known as magic methods or special methods, are a set of predefined methods in Python that have double underscores (__) at the beginning and end of their names. These methods are used to define the behavior of objects in various situations, such as when they are created, when they are printed, when they are compared, and so on. Dunder methods allow you to customize the behavior of your classes and make them more intuitive and user-friendly. Some common dunder methods include `__init__`, `__str__`, `__repr__`, `__add__`, `__eq__`, and many others. By implementing these methods in your classes, you can control how your objects behave in different contexts and make them more versatile and powerful.

It is tiresome to say under under method, so we call it dunder method, which is short for double underscore method.

#### Object Initialization

**`__init__`** : It is called when an object is created from a class. It is used to initialize the attributes
    of the object. The `__init__` method takes self as the first parameter, which refers to the instance of the
    class being created, and any additional parameters that are needed to initialize the object.

#### Object Representation

**`__repr__`** : It is used to define the “official” string representation of an object. It is meant to be unambiguous and is typically used for debugging and development. The goal of `__repr__` is to provide a string that, when passed to the eval() function, would recreate the original object (if possible). If this is not feasible, it should return a string that gives a detailed representation of the object.

**`__str__`** : It is used to define the “informal” or nicely printable string representation of an object. It is meant to be readable and is typically used for end-user display. The `__str__` method should return a string that is easy to read and provides a user-friendly representation of the object. If `__str__` is not defined, Python will fall back to using `__repr__` for the string representation of the object.

#### Iteration

**`__len__`** : It is used to define the behavior of the built-in `len()` function when called on an instance of the class. By implementing the `__len__` method, you can specify how the length of an object should be calculated and returned when `len()` is called on it. This allows you to make your custom objects compatible with the `len()` function, which can be useful for collections or any object that has a meaningful length.

**`__getitem__`** : It is used to define the behavior of the indexing operator ([]) when applied to an instance of the class. By implementing the `__getitem__` method, you can specify how the class should handle indexing operations, allowing you to access elements of the object using square brackets. This is particularly useful for classes that represent collections or sequences, as it enables you to retrieve specific items based on their index or key.

**`__reversed__`** : It is used to define the behavior of the built-in `reversed()` function when called on an instance of the class. By implementing the `__reversed__` method, you can specify how the class should handle reverse iteration, allowing you to iterate over the elements of the object in reverse order. This is particularly useful for classes that represent collections or sequences, as it enables you to easily access elements from the end to the beginning.

#### Comparison

**`__eq__`** : It is used to define the behavior of the equality operator (==) when comparing instances of the class. By implementing the `__eq__` method, you can specify how two instances of the class should be compared for equality, allowing you to determine if they are considered equal based on their attributes or other criteria. This is particularly useful for classes that represent data structures or entities where logical equality is important.

**`__lt__`** : It is used to define the behavior of the less-than operator (<) when comparing instances of the class. By implementing the `__lt__` method, you can specify how two instances of the class should be compared for ordering, allowing you to determine if one instance is considered less than another based on their attributes or other criteria. This is particularly useful for classes that represent sortable data structures or entities where ordering is important.

#### Arithmetic Operations

**`__add__`** : It is used to define the behavior of the addition operator (+) when applied to instances of the class. By implementing the `__add__` method, you can specify how two instances of the class should be added together, allowing you to define custom addition behavior based on their attributes or other criteria. This is particularly useful for classes that represent mathematical entities or data structures where addition is a meaningful operation.

#### Callable Objects

**`__call__`** : It is used to define the behavior of an instance of the class when it is called as a function. By implementing the `__call__` method, you can specify how the class should handle being called, allowing you to make instances of the class behave like functions. This is particularly useful for classes that represent callable entities or objects that need to perform a specific action when invoked.

#### Context Management

**`__enter__`** and **`__exit__`** : These methods are used to define the behavior of a class when it is used in a context manager (with statement). By implementing the `__enter__` and `__exit__` methods, you can specify how the class should handle entering and exiting a context, allowing you to manage resources or perform setup and teardown actions when using the class in a with statement. This is particularly useful for classes that represent resources that need to be properly managed, such as file handlers, database connections, or locks.

````python
class Account:
    """A simple account class"""
    
    """Object Initialization"""

    def __init__(self, owner, amount=0):
        """
        This is the constructor that lets us create
        objects from this class
        """
        self.owner = owner
        self.amount = amount
        self._transactions = []

    """Object Representation"""

    def __repr__(self):
        return 'Account({!r}, {!r})'.format(self.owner, self.amount)

    def __str__(self):
        return 'Account of {} with starting amount: {}'.format(
            self.owner, self.amount)
    

    """Iteration"""
    
    def add_transaction(self, amount):
        if not isinstance(amount, int):
            raise ValueError('please use int for amount')
        self._transactions.append(amount)

    @property
    def balance(self):
        return self.amount + sum(self._transactions)
    
    def __len__(self):
        return len(self._transactions)

    def __getitem__(self, position):
        return self._transactions[position]
    

    """Comparison"""
    
    def __eq__(self, other):
        return self.balance == other.balance

    def __lt__(self, other):
        return self.balance < other.balance
    

    """Addition"""

    def __add__(self, other):
        owner = '{}&{}'.format(self.owner, other.owner)
        start_amount = self.amount + other.amount
        acc = Account(owner, start_amount)
        for t in list(self) + list(other):
            acc.add_transaction(t)
        return acc
    
    """Call"""
    
    def __call__(self):
        print('Start amount: {}'.format(self.amount))
        print('Transactions: ')
        for transaction in self:
            print(transaction)
        print('\nBalance: {}'.format(self.balance))

    """Context Manager"""

    def __enter__(self):
        print('ENTER WITH: Making backup of transactions for rollback')
        self._copy_transactions = list(self._transactions)
        return self

    def __exit__(self, exc_type, exc_val, exc_tb):
        print('EXIT WITH:', end=' ')
        if exc_type:
            self._transactions = self._copy_transactions
            print('Rolling back to previous transactions')
            print('Transaction resulted in {} ({})'.format(
                exc_type.__name__, exc_val))
        else:
            print('Transaction OK')
````

````python
"""Object Representation Example"""

acc = Account('bob', 10)

print(str(acc)) #will return __str__ method output

print(acc) #will also return __str__ method output

print(repr(acc)) #will return __repr__ method output
````

#### Getter, Setter and Deleter

- `@property`: It is used to define a getter method for a property in a class. By using the `@property` decorator, you can create a method that can be accessed like an attribute, allowing you to retrieve the value of the property without explicitly calling a method. This is particularly useful for encapsulating data and providing controlled access to attributes of a class.

- `@<property_name>.setter`: It is used to define a setter method for a property in a class. By using the `@<property_name>.setter` decorator, you can create a method that allows you to set the value of the property while still providing controlled access to the attribute. This is useful for validating or processing the value before assigning it to the attribute.

- `@<property_name>.deleter`: It is used to define a deleter method for a property in a class. By using the `@<property_name>.deleter` decorator, you can create a method that allows you to delete the property while still providing controlled access to the attribute. This is useful for performing cleanup actions or releasing resources when the property is deleted.

````python
class Employee:

    num_of_emps = 0          # Always same for all the instances of class
    raise_amount = 1.04      # May differ from employee to employee

    def __init__(self, first_name, last_name, salary):
        self.first_name = first_name
        self.last_name = last_name
        self.salary = salary

        Employee.num_of_emps +=  1



    @property
    def email(self):
        return f"{self.first_name}.{self.last_name}@reversebits.com"

    @property
    def full_name(self):
        return f"{self.first_name} {self.last_name}"

    @full_name.setter
    def full_name(self, name):
        self.first_name, self.last_name = name.split(" ")

    @full_name.deleter
    def full_name(self):
        print("You just deleted the name of employee!!!")
        self.first_name = None
        self.last_name = None
````
