```python
#8 Create a function that takes a list of strings and returns the list sorted by the length of the strings

from stringInput import str_list

def sorted_list(input_list):
    input_list.sort()
    highest_len = 0
    
    for i in input_list:
        if len(i) > highest_len:
            highest_len = len(i)
            
    sorted_list = []
    for i in range(highest_len+1):
        for j in input_list:
            if len(j) == i:
                sorted_list.append(j)

    return sorted_list

input_list = str_list()
print(f"\nList : {input_list}\nSorted List : {sorted_list(input_list)}")
```

```python
#9 Write a program that checks if a given list is sorted in ascending order

from num import list_num

def is_sorted(input_list):
    test_list = input_list.copy()
    test_list.sort()

    if test_list == input_list:
        return True
    else:
        return False

input_list = list_num()
if is_sorted(input_list):
    print("\nThe list is sorted!!!")
else:
    print("\nSorry, The list is not sorted.")
```

```python
#13 Implement a function that removes a key-value pair from a dictionary

from num import list_num

def remove_element(input_dict, removal_key_list):
    try:
        for i in removal_key_list:
            input_dict.pop(i)
        return input_dict
    except Exception:
        print("Either the requested key is not present or dictionary is empty.")

input_dict = {1:'a', 2:'b', 3:'c', 4:'d', 5:'e'}

print(input_dict, "\n")
removal_key_list = list_num(input_msg="Enter the keys that you want to remove from the list saperated by comma for e.g. 2,3")

print(f"\nThe dictionary after requested removal of key value pairs : \n{remove_element(input_dict, removal_key_list)}")
```

```python
#15 Given a list of dictionaries, find the dictionary with the highest value for a specific key

input_list = [
    {'a':1, 'b':2, 'c':3, 'd':4},
    {'a':5, 'b':3, 'c':4, 'd':8},
    {'a':3, 'b':1, 'c':2, 'd':1}
]

print(input_list)
print()
key_input = input("Enter the key for which you want the maximum value : ")
list_key = []

try:
    for i in input_list:
        if i[key_input]:
            list_key.append(i[key_input])
            
except Exception:
    print("Either the requested key is not present or the dictionary is empty")

print(f"\nThe maximum value of {key_input} : {max(list_key)}")
```

```python
#16 Write a Python program that counts the number of occurrences of each character in a given string using a dictionary

string_input = input("Enter the string : ")
set_input = set(string_input)

count_dict = {i: 0 for i in set_input}

try:
    for i in string_input:
        count_dict[i] += 1
        
except Exception:
    print("Error Occured : There is some problem while counting the occurance")

print()
for c,o in count_dict.items():
    print(f"Character : {c} - Occurances : {o}")
```

```python
#19 Write a program that finds the average value of all the elements in a list of dictionaries

input_list = [
    {'a':1, 'b':2, 'c':3, 'd':4},
    {'a':5, 'b':3, 'c':4, 'd':8},
    {'a':3, 'b':1, 'c':2, 'd':1}
]

sum_terms = 0
terms = 0
for i in input_list:
    for j in i.values():
        sum_terms += j
        terms += 1

print(f"\nAverage of all the values in list of dictionary : {sum_terms/terms:.2f}")
```

```python
#20 Implement a function that takes a list of strings and returns a set of unique characters in all strings.

from stringInput import str_list

def unique_char(input_list):
    unique_set = set()

    for i in input_list:
        for j in i:
            unique_set.add(j)

    return unique_set

input_list = str_list()
print(f"\n The unique characters in given list are : {unique_char(input_list)}")
```
