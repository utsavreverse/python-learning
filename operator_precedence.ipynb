```python
"""Bitwise Operators"""

# Initial values for examples:
# a = 60 (binary: 0011 1100)
# b = 13 (binary: 0000 1101)
a = 60
b = 13

# --- 1. Basic Logic Operations ---
res_and = a & b        # AND: 12 (0000 1100) -> 1 if BOTH are 1
res_or  = a | b        # OR:  61 (0011 1101) -> 1 if EITHER is 1
res_xor = a ^ b        # XOR: 49 (0011 0001) -> 1 if DIFFERENT

# --- 2. The Tricky "NOT" Operator (~) ---
# Formula: ~x = -x - 1
res_not = ~a           # -61 (In C this would be a huge unsigned number like 4294967235)
                       # Why? Python uses infinite 2's complement bits.
                       # ...000000111100 (60) becomes ...111111000011 (-61)

# --- 3. Shift Operators ---
res_left  = a << 2     # Left Shift:  240 (1111 0000) -> Multiplies by 2^n
res_right = a >> 2     # Right Shift: 15  (0000 1111) -> Floor divides by 2^n

# --- 4. Compound Assignment ---
c = 10     # 1010
c &= 3     # c = c & 3 (0011) -> Result: 2 (0010)

# --- 5. Visualizing Bits (The bin() function) ---
print(bin(a))          # '0b111100'
print(f"{a:08b}")      # '00111100' (Format specifier for 8-bit padding)
```

```python
"""Logical Operators (Boolean Logic)"""

# Initial values
x = 10
y = 0
z = "Hello"

# --- 1. Basic Operators ---
res_and = (x > 5) and (x < 20)  # True (Both must be True)
res_or  = (x > 5) or (y > 5)    # True (At least one must be True)
res_not = not (x == 10)         # False (Inverts the boolean value)

# --- 2. Short-Circuit Evaluation (The "Lazy" Evaluation) ---
# Python stops evaluating as soon as the result is determined.

# OR: If the first value is Truthy, it returns the first value immediately.
# It does NOT check the second part (preventing the ZeroDivisionError).
safe_check = (x == 10) or (1 / 0)  
# Result: True (The error is never triggered!)

# AND: If the first value is Falsy, it returns the first value immediately.
safe_div = (y != 0) and (100 / y) 
# Result: False (or 0, depending on y). Stops before dividing by zero.

# --- 3. Returning Actual Values (Not just True/False) ---
# This is unique to Python and dynamic languages.
val1 = "Default" or "User Input"  # Returns "Default" (First Truthy value)
val2 = "" or "User Input"         # Returns "User Input" (First value was False, so return second)
val3 = "A" and "B"                # Returns "B" (First was True, so it kept going and returned the last checked value)

# --- 4. Chained Comparisons (Pythonic Syntactic Sugar) ---
# In C/Java:  if (age > 18 && age < 60)
# In Python:
age = 25
is_valid = 18 < age < 60          # True. Python expands this to: (18 < age) and (age < 60)

# --- 5. The "is" Operator (Identity vs Equality) ---
# '==' checks VALUES (Do they look the same?)
# 'is' checks MEMORY ADDRESS (Are they the same object?)
list_a = [1, 2, 3]
list_b = [1, 2, 3]

check_eq = (list_a == list_b)     # True  (Content is identical)
check_id = (list_a is list_b)     # False (Different memory locations)
```

### Operator Precedence

In python there are many types of operators like mathamatical(`+`, `-` etc), logical (`and`,`or` etc), bitwise (`&`,`|` etc), so python have specific rules that which operator should execute first for logical and correct output, and that execution order also known as **Operator precedence order**

```python
a = 5+10*2
print(a)
```

According to mathematical rules, the output of `5 + 10 * 2` is `25`. Python follows these rules: `10 * 2` is evaluated first because `*` (Multiplication) has higher precedence than `+` (Addition).

We can visualize this as `5 + (10 * 2)`. Since Parentheses `()` have the highest precedence in Python, they *force* the operation inside `(10 * 2)` to happen before it is added to `5`.

### Associativity

But what will happen if the expression has two or more operators with the same precedence? Well, in that case, Python chooses the execution order based on the **associativity** of those operators.

```python
# '-' and '+' operators have same precedence, and they are left associative
# so 3-5 will be evaluated first and then result will be added with 6
b = 3-5+6
print(b)

# '**' is right associative so 3**2 evaluated first and then result will go in power of 4
c = 4**3**2
print(c)
```

```python
"""Tricky right assosiativity of conditional statement"""

def prover(x):
    print(f"Testing with x = {x}")
    
    # 1. Standard Python Behavior (Implicit Right Associativity)
    # Logic: Is it even? If yes, stop. If no, check if big.
    res_standard = "Even" if x % 2 == 0 else "Big" if x > 10 else "Other"
    print(f"Standard (Right): {res_standard}")

    # 2. Explicit Right Associativity (Same as Standard)
    # Logic: "Even" if Even else ("Big" if > 10 else "Other")
    res_right = "Even" if x % 2 == 0 else ("Big" if x > 10 else "Other")
    print(f"Explicit Right:   {res_right}")

    # 3. Forced Left Associativity (Completely Different Logic)
    # Logic: Decide ("Even" or "Big") first. THEN check if x > 10 to keep it.
    res_left = ("Even" if x % 2 == 0 else "Big") if x > 10 else "Other"
    print(f"Forced Left:      {res_left}")

# Run the proof
prover(4)
```

| Operator | Description | Associativity |
| :--- | :--- | :--- |

| `x[i]`, `x[i:j]`, `x(...)`, `x.attr` | **Access & Call**<br>Subscription, slicing, function calls, attribute references. | Left-to-right |

| `**` | **Exponentiation**<br>Raises a number to the power of another. | **Right-to-left** |
| `+x`, `-x`, `~x` | **Unary Arithmetic & Bitwise**<br>Positive, negative, and bitwise NOT (inversion). | N/A (Unary) |
| `*`, `@`, `/`, `//`, `%` | **Multiplicative**<br>Multiplication, matrix multiplication (`@`), division, floor division, remainder (modulo). | Left-to-right |
| `+`, `-` | **Additive**<br>Addition and subtraction. | Left-to-right |
| `<<`, `>>` | **Shifts**<br>Bitwise left shift and bitwise right shift. | Left-to-right |
| `&` | **Bitwise AND**<br>Logical AND on individual bits. | Left-to-right |
| `^` | **Bitwise XOR**<br>Logical exclusive OR on individual bits. | Left-to-right |
| `\|` | **Bitwise OR**<br>Logical OR on individual bits. | Left-to-right |
| `in`, `not in`, `is`, `is not`, `<`, `<=`, `>`, `>=`, `!=`, `==` | **Comparisons**<br>Membership tests, identity tests, and value comparisons. | **Chained**<br>*(e.g., `a < b < c`)* |
| `not x` | **Boolean NOT**<br>Logical negation. | N/A (Unary) |
| `and` | **Boolean AND**<br>Logical AND (short-circuiting). | Left-to-right |
| `or` | **Boolean OR**<br>Logical OR (short-circuiting). | Left-to-right |
| `if` – `else` | **Conditional Expression**<br>Ternary operator (e.g., `x if C else y`). | **Right-to-left** |
| `lambda` | **Lambda**<br>Anonymous function definition. | N/A |
| `:=` | **Assignment Expression**<br>The "walrus" operator (assigns a value to a variable within an expression). | Right-to-left<br>*(Requires parentheses to nest)* |
