```python
#1 Create a class representing a Student with attributes like name, age, and grades.

class Student:
    def __init__(self, name, age, grades):
        self.name = name
        self.age = age
        self.grades = grades

    def average_grade(self):
        return sum(self.grades) / len(self.grades)

student1 = Student("Meow", 20, [85, 90, 78])
print(f"Student Name: {student1.name}")
print(f"Student Age: {student1.age}")
print(f"Student Grades: {student1.grades}")
print(f"Average Grade: {student1.average_grade():.2f}")
```

```python
#3 Implement a program that simulates a basic bank account using a BankAccount class.

class BalanceNotAvailable(Exception):
    pass

class BankAccount:
    __password = "password"
    __balance = 10000

    def __init__(self):
        pass

    def deposite(self, amount):
        password_input = input("Please enter your password : ")
        if (password_input == self.__password):
            try:
                if amount > 0:
                    self.__balance += amount
                    print("\nAmount deposited sucessfully!")
                    print(f"\nYour Account balance is {self.__balance} rupees.")
                else:
                    raise ValueError
            except ValueError:
                print("\nERROR : PLEASE ENTER THE VALID AMOUNT YOU WANT TO DEPOSITE")
        else:
            print("\n You entered the wrong password, please try again.")

    def withdraw(self, amount):
        password_input = input("Please enter your password : ")
        if (password_input == self.__password):
            try:
                if amount > self.__balance:
                    raise BalanceNotAvailable
                else:
                    if amount > 0:
                        self.__balance -= amount
                        print("\nAmount withdrawed sucessfully!")
                        print(f"\nYour Account balance is {self.__balance} rupees.")
                    else:
                        raise ValueError
            except BalanceNotAvailable:
                print("\nERROR : YOU DON'T HAVE THE SUFFICIENT FUNDS")
            except ValueError:
                print("\nERROR : PLEASE ENTER THE VALID AMOUNT YOU WANT TO DEPOSITE")
        else:
            print("\n You entered the wrong password, please try again.")

    def check_balance(self):
        password_input = input("Please enter your password : ")
        if (password_input == self.__password):
            print(f"\nYour Account balance is {self.__balance} rupees.")
        else:
            print("\n You entered the wrong password, please try again.")

    def password_reset(self):
        old_pass = input("Please enter your current password : ")
        if (old_pass == self.__password):
            while True:
                new_pass = input("Please enter your new password : ")
                confirm_new_pass = input("Please re-enter your new password : ")
                if new_pass == confirm_new_pass and new_pass != old_pass:
                    self.__password = new_pass
                    print("\nYour password changed sucessfully!")
                    break
                else:
                    if new_pass != confirm_new_pass:
                        print("Your does'nt entered the same passwords for new password prompts, please try again.")
                    if new_pass == old_pass:
                        print("New password can't be same as the old password, please try again.")
        else:
            print("\n You entered the wrong password, please try again.")


user1 = BankAccount()

while True:
    print("""
------------- Welcome to the Bank ---------------
1. Deposite
2. Withdraw
3. Check Balance
4. Change Password
0. Quit""")
    choice = int(input("\nPlease enter your choice : "))

    match choice:
        case 1:
            amount = int(input("\nEnter the amount you want to deposite : "))
            user1.deposite(amount)
        case 2:
            amount = int(input("\nEnter the amount you want to withdraw : "))
            user1.withdraw(amount)
        case 3:
            user1.check_balance()
        case 4:
            user1.password_reset()
        case 0:
            print("Thank you for using the bank service.")
            break
        case _:
            print("\nPlease enter a valid input.")
```

```python
# 11 Create a base class Shape with methods to calculate area and perimeter and derive classes Circle and Square.

from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius

class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side ** 2

    def perimeter(self):
        return 4 * self.side

radius = float(input("Enter the radius of the circle : "))
circle1 = Circle(radius)
print(f"\nThe area of the circle is {circle1.area():.2f} square units.")
print(f"The perimeter of the circle is {circle1.perimeter():.2f} units.")

side = float(input("\nEnter the side of the square : "))
square1 = Square(side)
print(f"\nThe area of the square is {square1.area():.2f} square units.")
print(f"The perimeter of the square is {square1.perimeter():.2f} units.")
```

```python
# 12 Implement a class hierarchy to represent different types of employees (Manager, Engineer) with their attributes.

class Employee:
    def __init__(self, name, age, salary):
        self.name = name
        self.age = age
        self.salary = salary

    def display_info(self):
        print(f"\nEmployee Information : \nName : {self.name}\nAge : {self.age}\nSalary : {self.salary}")

class Manager(Employee):
    def __init__(self, name, age, salary, department):
        super().__init__(name, age, salary)
        self.department = department

    def display_info(self):
        super().display_info()
        print(f"Department : {self.department}")

class Engineer(Employee):
    def __init__(self, name, age, salary, specialization):
        super().__init__(name, age, salary)
        self.specialization = specialization

    def display_info(self):
        super().display_info()
        print(f"Specialization : {self.specialization}")

name = input("Enter the name of the manager : ")
age = int(input("Enter the age of the manager : "))
salary = float(input("Enter the salary of the manager : "))
department = input("Enter the department of the manager : ")
manager1 = Manager(name, age, salary, department)
manager1.display_info()

name = input("\nEnter the name of the engineer : ")
age = int(input("Enter the age of the engineer : "))
salary = float(input("Enter the salary of the engineer : "))
specialization = input("Enter the specialization of the engineer : ")
engineer1 = Engineer(name, age, salary, specialization)
engineer1.display_info()
```

```python
#13 Write a Python program that uses inheritance to represent a hierarchy of shapes (Triangle, Rectangle, etc.)

class Shape:
    def area(self):
        pass

    def perimeter(self):
        pass

class Triangle(Shape):
    def __init__(self, base, height, side1, side2, side3):
        self.base = base
        self.height = height
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3

    def area(self):
        return 0.5 * self.base * self.height

    def perimeter(self):
        return self.side1 + self.side2 + self.side3

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

base = float(input("Enter the base of the triangle : "))
height = float(input("Enter the height of the triangle : "))
side1 = float(input("Enter the length of side 1 of the triangle : "))
side2 = float(input("Enter the length of side 2 of the triangle : "))
side3 = float(input("Enter the length of side 3 of the triangle : "))
triangle1 = Triangle(base, height, side1, side2, side3)
print(f"\nThe area of the triangle is {triangle1.area():.2f} square units.")
print(f"The perimeter of the triangle is {triangle1.perimeter():.2f} units.")

length = float(input("\nEnter the length of the rectangle : "))
width = float(input("Enter the width of the rectangle : "))
rectangle1 = Rectangle(length, width)
print(f"\nThe area of the rectangle is {rectangle1.area():.2f} square units.")
print(f"The perimeter of the rectangle is {rectangle1.perimeter():.2f} units.")
```

```python
#14 Create a class hierarchy to represent different types of animals (Birds, Fish) with their attributes and methods.

class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

    def display_info(self):
        print(f"\nAnimal Information : \nName : {self.name}\nSpecies : {self.species}")

class Bird(Animal):
    def __init__(self, name, species, wing_len):
        super().__init__(name, species)
        self.wing_len = wing_len

    def display_info(self):
        super().display_info()
        print(f"Wing Length : {self.wing_len} meters")

class Fish(Animal):
    def __init__(self, name, species, water_type):
        super().__init__(name, species)
        self.water_type = water_type

    def display_info(self):
        super().display_info()
        print(f"Water Type : {self.water_type}")

name = input("Enter the name of the bird : ")
species = input("Enter the species of the bird : ")
wing_len = float(input("Enter the wing length of the bird in meters : "))
bird1 = Bird(name, species, wing_len)
bird1.display_info()

name = input("\nEnter the name of the fish : ")
species = input("Enter the species of the fish : ")
water_type = input("Enter the water type of the fish (normal / sea) : ")
fish1 = Fish(name, species, water_type)
fish1.display_info()
```

```python
#16 Implement a program that uses inheritance to represent a hierarchy of vehicles (Car, Bike, Truck, etc.)

class Vehicle:
    def __init__(self, maker, model, year):
        self.maker = maker
        self.model = model
        self.year = year

    def display_info(self):
        print(f"\nVehicle Information : \nMake : {self.maker}\nModel : {self.model}\nYear : {self.year}")

class Car(Vehicle):
    def __init__(self, maker, model, year, num_doors):
        super().__init__(maker, model, year)
        self.num_doors = num_doors

    def display_info(self):
        super().display_info()
        print(f"Number of Doors : {self.num_doors}")

class Bike(Vehicle):
    def __init__(self, maker, model, year, type_bike):
        super().__init__(maker, model, year)
        self.type_bike = type_bike

    def display_info(self):
        super().display_info()
        print(f"Type of Bike : {self.type_bike}")

class Truck(Vehicle):
    def __init__(self, maker, model, year, capacity):
        super().__init__(maker, model, year)
        self.capacity = capacity

    def display_info(self):
        super().display_info()
        print(f"Capacity : {self.capacity} tons")

maker = input("Enter the maker of the car : ")
model = input("Enter the model of the car : ")
year = int(input("Enter the year of the car : "))
num_doors = int(input("Enter the number of doors of the car : "))
car1 = Car(maker, model, year, num_doors)
car1.display_info()

maker = input("\nEnter the maker of the bike : ")
model = input("Enter the model of the bike : ")
year = int(input("Enter the year of the bike : "))
type_bike = input("Enter the type of the bike : ")
bike1 = Bike(maker, model, year, type_bike)
bike1.display_info()

maker = input("\nEnter the maker of the truck : ")
model = input("Enter the model of the truck : ")
year = int(input("Enter the year of the truck : "))
capacity = float(input("Enter the capacity of the truck in tons : "))
truck1 = Truck(maker, model, year, capacity)
truck1.display_info()
```

```python
#17 Write a Python program that uses encapsulation to protect sensitive information in a User class.

class User:
    def __init__(self, username, password):
        self.username = username
        self.__password = password

    def display_info(self):
        print(f"\nUser Information : \nUsername : {self.username}")

    def check_password(self, password):
        return self.__password == password

username = input("Enter the username : ")
password = input("Enter the password : ")
user1 = User(username, password)
user1.display_info()
password_input = input("Please enter your password to check : ")
if user1.check_password(password_input):
    print("Password is correct!")
else:
    print("Password is incorrect!")
```
