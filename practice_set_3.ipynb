```python
#3 Implement a program that reads a text file and counts the number of words and lines in it
import re

lines_no = 0
words_no = 0

with open("quotes.txt","r") as rf:
    for line in rf:
        words = re.findall(r'\S+', line)
        words_no += len(words)
        lines_no += 1

print(f"""
Number of lines - {lines_no}
Number of words - {words_no}""")
```

```python
#6 Write a program that reads a CSV file and finds the total sales revenue for a specific product

import numpy as np
import pandas as pd
import num

df = pd.read_csv("products.csv")
product_list = np.array(df["Product Name"])
selling_list = np.array(df["Selling Price"])
unique_products = np.unique(df["Product Name"])
input_dict = {key+1 : unique_products[key] for key in range(0,len(unique_products))}

print("------------- List of items -------------------")
for key,val in input_dict.items():
    print(f"{key} - {val}")
print("-----------------------------------------------")
input_option = num.list_num_positive(input_msg="Enter the respective number crossponding to the product for which you want to see the revenue : ",err_msg="Please enter the valid number")
total_sales = 0

for i in range(0,len(product_list)):
    if product_list[i] == input_dict[input_option]:
        total_sales += selling_list[i]

print(f"\n The total sales revenue of {input_dict[input_option]} = {total_sales:.2f}")
```

```python
#7 Given a text file with a list of numbers, write a function that finds the sum of all numbers in the file
import re

def find_sum():
    total_sum = 0
    with open("numbers_ex.txt","r") as rf:
        for line in rf:
            words = re.findall(r'\S+', line)
            words = [wd.replace(",","") for wd in words]
            nums = []
            for word in words:
                if str.isnumeric(word):
                    nums.append(int(word))
            total_sum += sum(nums)
    return total_sum

print(f"\nThe total sum of all the numbers in file is = {find_sum()}")
```

```python
#9 Write a function that reads a JSON file and extracts specific information from it

import json

with open("orders.json","r") as of:
    order_data = json.load(of)

selling_total = 0
for i in order_data["customers"]:
    for j in i["orders"]:
        for k in j["items"]:
            selling_total += k["price"]

print(f"\nThe total revenue for shop currently is {selling_total}")
```
