# String Formatting in Python

In real-world Python programming, you often need to insert variables or objects directly into strings. While simple concatenation is an option, string formatting is a much more efficient and readable solution. This approach utilizes **String interpolation** and Python's Format Specification **Mini-Language**, which are discussed below.

### String interpolation

String interpolation is a method for generating strings by inserting variables or objects at specific locations within a base string. Often, simply inserting the value isn't enough; you may also need to format it (e.g., setting decimal precision, padding with spaces) to produce a clean, readable result.

Python provides three main tools that support string formatting and interpolation:

1. F-strings (formatted string literals)
2. The `str.format()` method
3. The modulo operator (%)

The first two methods (F-strings and `str.format()`) support Python's Format Specification Mini-Language, offering powerful styling options. The third tool (the % operator) is older, has fewer formatting capabilities, and is rarely used in modern codebases for general string manipulation.

F-strings are the modern standard and appear most frequently in new code. However, `str.format()` retains a distinct advantage: it can function as a reusable template. Unlike F-strings, which evaluate immediately, a format string can be defined once and applied to multiple different data instances later.

```python
#Example of string interpolation using F-string
firstHalf = "Hello"
print(f"{firstHalf} World!")
```

### Format Specification Mini-Language

The Format Specification Mini-Language consists of optional parameters that define how a value will appear after formatting. There are many useful ways to utilize these options, some of which are covered in this notebook. For a complete reference, you can consult the official [Python Docs](https://docs.python.org/3/library/string.html#format-specification-mini-language).

## F-string (Formatted string literals)

F-strings are the latest major method for inserting variables or objects into strings. Introduced in Python 3.6, they are now heavily used by programmers because they are more readable and concise than older methods.

```python
"""F-string"""

import datetime

# --- Setup Variables ---
name = "Meow"
age = 9
price = 1234.5678
percent = 0.25
d = datetime.datetime(2025, 1, 27, 15, 30)

# --- 1. Basic Interpolation ---
# Just inserting variables directly
basic = f"Hello, {name}. You are {age} years old."

# --- 2. Expressions & Function Calls ---
# You can run code inside the curly braces
calc = f"Next year, you will be {age + 1}."
func = f"Your name in lowercase is '{name.lower()}'."

# --- 3. Format Specification Mini-Language ---
# Controlling how numbers look (Precision, Commas, Percentages)
formatted_float = f"Price: ${price:.2f}"        # Fixed point: 1234.57
formatted_comma = f"Price: ${price:,.2f}"       # Comma separator: 1,234.57
formatted_pct   = f"Tax: {percent:.1%}"         # Percentage: 25.0%

# --- 4. Alignment and Padding ---
# Structure: {value : fill align width}
# < = Left, > = Right, ^ = Center
align_right  = f"|{name:>10}|"   # Pad with spaces to 10 chars, right align
align_center = f"|{name:^10}|"   # Pad with spaces to 10 chars, center align
pad_zeros    = f"ID: {age:05d}"  # Pad with zeros to 5 digits (00009)
pad_custom   = f"|{name:_<10}|"  # Pad with underscores, left align

# --- 5. Date Formatting ---
# F-strings can interpret datetime objects directly
date_str = f"Meeting is on: {d:%B %d, %Y at %H:%M}"

# --- 6. Debugging (Python 3.8+) ---
# Using '=' prints the variable name AND the value. Very useful for logging.
debug_val = f"{name=}"    # Output: "name='Alice'"
debug_calc = f"{age+5=}"  # Output: "age+5=35"

# --- 7. Escaping Braces ---
# Use double braces {{ }} to show literal curly braces
escaped = f"This is a set: {{1, 2, 3}} and this is a variable: {age}"

# --- Printing Results ---
print(f"1. Basic:       {basic}")
print(f"2. Calculation: {calc}")
print(f"3. Formatting:  {formatted_comma} | {formatted_pct}")
print(f"4. Alignment:   {align_right} | {align_center} | {pad_zeros} | {pad_custom}")
print(f"5. Date:        {date_str}")
print(f"6. Debugging:   {debug_val} | {debug_calc}")
print(f"7. Escaping:    {escaped}")
```
