# Slicing

Before we dive into slicing, let's understand that how indexing works in slicing operations.

slicing syntax: `list [ start : stop : step]`

- `start`: The index where the slice starts (inclusive). Default is 0.
- `stop`: The index where the slice ends (exclusive). Default is the length of the list.
- `step`: The interval between each index in the slice. Default is 1.

Now understand that `start` and `stop` are just indices, and they can be positive or negative, whereas `step` determines the stride of the slice, meaning how many elements to skip between each index, it can also be negative which indicates slicing in reverse order.

In slicing, the `stop` index is exclusive, meaning the element at the `stop` index is not included in the slice. This is a common convention in programming and helps to avoid off-by-one errors, you can understand this better with examples below.

```python
"""

 +---+---+---+---+---+---+
 | P | y | t | h | o | n |
 +---+---+---+---+---+---+
 0   1   2   3   4   5   6
-6  -5  -4  -3  -2  -1

"""
```

```python
"""Positive indexing examples"""

text = "Python"

# Accessing first character
first_char = text[0]  # 'P'
# Accessing last character
last_char = text[-1]  # 'n'
# Slicing from index 1 to 4
slice_1_4 = text[1:4]  # 'yth'
# Slicing from start to index 3
slice_start_3 = text[:3]  # 'Pyt'
# Slicing from index 3 to end
slice_3_end = text[3:]  # 'hon' 
# Slicing with step
slice_step_2 = text[::2]  # 'Pto'
# Reversing the string
reversed_text = text[::-1]  # 'nohtyP'

"""Negative indexing examples"""
text = "Python"
# Accessing first character
first_char = text[-6]  # 'P'
# Accessing last character
last_char = text[5]  # 'n'
# Slicing from index -5 to -2
slice_neg5_neg2 = text[-5:-2]  # 'yth'
# Slicing from start to index -3
slice_start_neg3 = text[:-3]  # 'Pyt'
# Slicing from index -3 to end
slice_neg3_end = text[-3:]  # 'hon'
# Slicing with step
slice_step_2 = text[::-2]  # 'nhy'
# Reversing the string
reversed_text = text[::-1]  # 'nohtyP'
```

# Comprehensions

List comprehensions (it can work with any iterable) provide a concise way to create lists. Common applications are to make new lists where each element is the result of some operations applied to each member of another sequence or iterable, or to create a subsequence of those elements that satisfy a certain condition.

### List comprehension

__new_list = \[ expression for member in iterable \]__

```python
squares = []

for num in range(1, 11):
    squares.append(num*num)

print(squares)

del squares

squares = [num*num for num in range(1, 11)]
print(squares)
```

with conditions

1. __new_list = \[expression for member in iterable (if conditional)\]__  => filter items
2. __new_list = \[expression (if conditional) for member in iterable\]__  => change items

```python
# Get all the numbers between 1 to 1000 whose square is less than 50000.

squares_lt_five_lac = [num for num in range(1, 1001) if num*num < 50000]

print(squares_lt_five_lac)

# only those elements will be added to new list for which the condition is True.
```

```python
# Get all the vowels in a sentence.

sentence = "Well, Python is different than others!"

vowels = "aeiou"

vowels_in_sentence = [char for char in sentence if char in vowels]

print(vowels_in_sentence)
```

```python
# get squares of even numbers and cubes of odd numbers in a range.

squares_and_cubes = [i*i*i if i%2 else i*i for i in range(1, 21)]
print(squares_and_cubes)
```

### Set comprehension

```python
# find unique vowels from a sentence.

sentence = "Get me all the unique vowels man!"

vowels = "aeiou"

unique_vowels_in_sentence = {char for char in sentence if char in vowels}

print(unique_vowels_in_sentence)
```

### Dictionary comprehension

```python
# creat a dict out of list of numbers with number as a key and their square roots as value.

import math

squares = {num: math.sqrt(num) for num in range(1, 11)}
print(squares)
```
